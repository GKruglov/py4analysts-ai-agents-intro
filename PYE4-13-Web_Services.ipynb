## **Использование веб-сервисов**

После того как появилась возможность программно получать и разбирать документы по протоколу HTTP, довольно быстро появился подход — создавать документы, специально предназначенные для обработки другими программами (то есть не HTML для отображения в браузере).

Существует два распространённых формата, которые используются при обмене данными через Интернет. Расширяемый язык разметки (eXtensible Markup Language, XML) применяется уже очень давно и лучше всего подходит для обмена данными документного типа. Когда программам требуется обмениваться словарями, списками и другой внутренней информацией между собой, используется формат <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">JSON</span> (JavaScript Object Notation, см. [www.json.org](http://www.json.org)). Ниже рассмотрены оба формата.


### **Расширяемый язык разметки — XML**

XML внешне очень похож на HTML, но имеет более строгую структуру. Ниже приведён пример XML-документа:

```xml
<person>
  <name>Chuck</name>
  <phone type="intl">
    +1 734 303 4456
  </phone>
  <email hide="yes" />
</person>
```

Каждая пара открывающего (например, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">&lt;person&gt;</span>) и закрывающего тегов (например, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">&lt;/person&gt;</span>) представляет собой **элемент** или **узел** с тем же именем, что и тег (например, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">person</span>). Элемент может содержать текст, атрибуты (например, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">hide</span>) и другие вложенные элементы. Если XML-элемент пуст (то есть не содержит контента), он может быть записан в виде самозакрывающегося тега (например, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">&lt;email /&gt;</span>).

Часто XML-документ удобно представлять как древовидную структуру, в которой есть корневой элемент (в данном случае <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">person</span>), а другие теги (например, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">phone</span>) являются **потомками** своих **родительских** элементов.

### **Разбор XML**

Ниже приведено простое приложение, которое разбирает XML и извлекает из него некоторые элементы данных:


````python
import xml.etree.ElementTree as ET

data = '''
<person>
  <name>Chuck</name>
  <phone type="intl">
    +1 734 303 4456
  </phone>
  <email hide="yes" />
</person>'''

tree = ET.fromstring(data)
print('Name:', tree.find('name').text)
print('Attr:', tree.find('email').get('hide'))
````

Тройные одинарные кавычки (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">&apos;&apos;&apos;</span>), как и тройные двойные (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">&quot;&quot;&quot;</span>), позволяют создавать строки, занимающие несколько строк.

Вызов <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">fromstring</span> преобразует строковое представление XML в «дерево» XML-элементов. Когда XML находится в виде дерева, доступен ряд методов для извлечения данных из него. Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span> выполняет поиск в дереве XML и возвращает элемент, соответствующий указанному тегу.

Использование XML-парсера, такого как <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">ElementTree</span>, имеет преимущество: хотя XML в данном примере довольно прост, существует множество правил, касающихся корректности XML. <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">ElementTree</span> позволяет извлекать данные, не беспокоясь о тонкостях синтаксиса XML.


### **Обход узлов в цикле**

Нередко XML содержит несколько узлов, и требуется написать цикл для их обработки. В следующей программе выполняется обход всех узлов <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">user</span>:


````python
import xml.etree.ElementTree as ET

input = '''
<stuff>
  <users>
    <user x="2">
      <id>001</id>
      <name>Chuck</name>
    </user>
    <user x="7">
      <id>009</id>
      <name>Brent</name>
    </user>
  </users>
</stuff>'''

stuff = ET.fromstring(input)
lst = stuff.findall('users/user')
print('User count:', len(lst))

for item in lst:
    print('Name', item.find('name').text)
    print('Id', item.find('id').text)
    print('Attribute', item.get('x'))
````

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">findall</span> возвращает список поддеревьев Python, соответствующих структурам <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">user</span> в XML-дереве. Затем можно написать цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>, который проходит по каждому узлу пользователя и выводит текстовые элементы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">name</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">id</span>, а также атрибут <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> узла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">user</span>.

Важно указывать в <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">findall</span> все промежуточные уровни вложенности, кроме корневого элемента (например, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">users/user</span>). В противном случае Python не найдёт нужных узлов.


````python
import xml.etree.ElementTree as ET

input = '''
<stuff>
  <users>
    <user x="2">
      <id>001</id>
      <name>Chuck</name>
    </user>
    <user x="7">
      <id>009</id>
      <name>Brent</name>
    </user>
  </users>
</stuff>'''

stuff = ET.fromstring(input)

lst = stuff.findall('users/user')
print('User count:', len(lst))

lst2 = stuff.findall('user')
print('User count:', len(lst2))
````

Переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">lst</span> содержит все элементы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">user</span>, вложенные в родительский элемент <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">users</span>. Переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">lst2</span> ищет элементы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">user</span>, непосредственно вложенные в корневой элемент <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">stuff</span>, — а таких там нет.


### **JavaScript Object Notation — JSON**

Формат JSON был вдохновлён синтаксисом объектов и массивов языка JavaScript. Однако, поскольку Python был создан раньше JavaScript, синтаксис словарей и списков Python повлиял на синтаксис JSON. Поэтому JSON практически идентичен комбинации списков и словарей Python.

Ниже представлено JSON-представление, приблизительно эквивалентное простому XML-примеру выше:

```json
{
  "name" : "Chuck",
  "phone" : {
    "type" : "intl",
    "number" : "+1 734 303 4456"
  },
  "email" : {
    "hide" : "yes"
  }
}
```

Заметны некоторые различия. Во-первых, в XML к тегу «phone» можно добавлять атрибуты, например «intl». В JSON используются только пары ключ-значение. Также тег «person» из XML отсутствует — его заменяет пара внешних фигурных скобок.

В целом, структуры JSON проще, чем XML, поскольку JSON обладает меньшим количеством возможностей. Однако преимущество JSON в том, что он **напрямую** отображается на комбинацию словарей и списков. Поскольку в почти всех языках программирования есть аналоги словарей и списков Python, JSON является очень удобным форматом для обмена данными между двумя взаимодействующими программами.

JSON быстро становится предпочтительным форматом для обмена данными между приложениями благодаря своей относительной простоте по сравнению с XML.


### **Разбор JSON**

JSON-структуры формируются путём вложения словарей и списков в нужной последовательности. В следующем примере мы представляем список пользователей, где каждый пользователь — это набор пар ключ-значение (то есть словарь). Таким образом, получается список словарей.

В следующей программе используется встроенная библиотека <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">json</span> для разбора JSON и чтения данных. Сравните её с аналогичным кодом для XML выше. JSON содержит меньше деталей, поэтому необходимо заранее знать, что ожидается список пользователей и что каждый пользователь — это набор пар ключ-значение. JSON более лаконичен (преимущество), но и менее самоописателен (недостаток).


````python
import json

data = '''
[
  { "id" : "001",
    "x" : "2",
    "name" : "Chuck"
  } ,
  { "id" : "009",
    "x" : "7",
    "name" : "Brent"
  }
]'''

info = json.loads(data)
print('User count:', len(info))

for item in info:
    print('Name', item['name'])
    print('Id', item['id'])
    print('Attribute', item['x'])
````

Если сравнить код извлечения данных из разобранных JSON и XML, можно увидеть, что <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">json.loads()</span> возвращает список Python, по которому осуществляется обход с помощью цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>. Каждый элемент этого списка — словарь Python. После разбора JSON можно использовать обычный оператор индекса для извлечения различных фрагментов данных о каждом пользователе. Нет необходимости применять библиотеку JSON для «погружения» в разобранные данные, поскольку возвращённые данные — это уже родные структуры Python.

Вывод этой программы полностью совпадает с выводом XML-версии выше.

В индустрии наблюдается общий тренд перехода от XML к JSON в веб-сервисах. Поскольку JSON проще и напрямую соответствует нативным структурам данных, код разбора и извлечения, как правило, становится проще и прямолинейнее. Тем не менее, XML является более самоописательным, поэтому в некоторых приложениях он сохраняет своё преимущество. Например, большинство текстовых процессоров хранит документы во внутреннем XML-формате, а не в JSON.


### **Программные интерфейсы приложений**

Теперь у нас есть возможность обмениваться данными между приложениями по протоколу HTTP, а также способ представлять сложные данные, передаваемые между приложениями, — с использованием XML или JSON.

Следующий шаг — начать определять и документировать «контракты» между приложениями, использующими эти техники. Общее название для таких соглашений между приложениями — **программные интерфейсы приложений** (Application Program Interfaces, API). Когда используется API, одна программа обычно предоставляет набор **сервисов** для использования другими приложениями и публикует API (то есть «правила»), которым необходимо следовать для обращения к этим сервисам.

Когда мы начинаем строить программы, функциональность которых включает обращение к сервисам, предоставляемым другими программами, такой подход называется **сервис-ориентированной архитектурой** (Service-Oriented Architecture, SOA). При SOA-подходе наше приложение использует сервисы других приложений. Противоположный (не-SOA) подход — это отдельное самостоятельное приложение, содержащее весь код, необходимый для его работы.

Множество примеров SOA можно увидеть при работе с Интернетом. Можно зайти на один веб-сайт и забронировать авиабилет, отель и автомобиль из одного места. Данные об отелях хранятся не на компьютерах авиакомпании. Компьютеры авиакомпании обращаются к сервисам на серверах отелей, получают данные и представляют их пользователю. Когда пользователь соглашается на бронирование, сайт авиакомпании использует другой веб-сервис отеля, чтобы фактически оформить бронирование. Когда доходит до оплаты кредитной картой, в процесс вовлекаются ещё другие компьютеры.

![Сервис-ориентированная архитектура](./PYE4-13-Web_Services_files/soa.svg)

*Сервис-ориентированная архитектура*

Преимущества сервис-ориентированной архитектуры многочисленны: (1) данные всегда хранятся в единственном экземпляре (это особенно важно, например, для бронирования отелей, чтобы не допустить двойной продажи); (2) владельцы данных могут устанавливать правила их использования. Однако, чтобы обеспечить хорошую производительность и удовлетворить потребности пользователя, SOA-система должна быть тщательно спроектирована.

Когда приложение предоставляет набор сервисов своего API через Интернет, такие сервисы называются **веб-сервисами**.


### **Безопасность и использование API**

Достаточно часто для использования API поставщика требуется API-ключ. Основная идея заключается в том, что поставщик хочет знать, кто использует его сервисы и в каком объёме. Возможно, у него есть бесплатные и платные тарифы либо политика ограничения количества запросов от одного пользователя за определённый период времени.

Иногда после получения API-ключа его достаточно просто включить в POST-данные или передать как параметр URL при обращении к API.

В других случаях поставщик требует более надёжной проверки источника запросов и ожидает получения криптографически подписанных сообщений с использованием общих ключей и секретов. Очень распространённая технология подписи запросов через Интернет называется **OAuth**. Подробнее о протоколе OAuth можно прочитать на сайте [www.oauth.net](http://www.oauth.net).

К счастью, существует множество удобных и бесплатных OAuth-библиотек, поэтому нет необходимости реализовывать OAuth с нуля, опираясь только на спецификацию. Эти библиотеки имеют различную степень сложности и функциональности. На сайте OAuth доступна информация о различных библиотеках.


### **Глоссарий**

**API** (Application Program Interface, программный интерфейс приложения) — контракт между приложениями, определяющий закономерности, по которым данные передаются между ними.

**ElementTree** — встроенная библиотека Python, используемая для разбора XML-данных.

**JSON** (JavaScript Object Notation) — формат, позволяющий кодировать структурированные данные на основе синтаксиса объектов JavaScript.

**SOA** (Service-Oriented Architecture, сервис-ориентированная архитектура) — подход, при котором приложение состоит из компонентов, соединённых через сеть.

**XML** (eXtensible Markup Language, расширяемый язык разметки) — формат, позволяющий кодировать структурированные данные.
