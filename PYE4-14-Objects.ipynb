### **Использование объектов**

Как оказывается, мы уже использовали объекты на протяжении всей книги. Python предоставляет множество встроенных объектов. Ниже приведён простой код, в котором первые несколько строк должны выглядеть естественно и знакомо.


````python
stuff = list()
stuff.append('python')
stuff.append('chuck')
stuff.sort()
print (stuff[0])
print (stuff.__getitem__(0))
print (list.__getitem__(stuff,0))
````

Не сосредотачиваясь на том, что делает этот код, посмотрим, что в действительности происходит с точки зрения объектно-ориентированного программирования. Не переживайте, если следующие абзацы не будут понятны при первом чтении, — мы ещё не определили все используемые термины.

Первая строка **конструирует** объект типа <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">list</span>; вторая и третья строки **вызывают** метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">append()</span>; четвёртая строка вызывает метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sort()</span>; пятая строка **извлекает** элемент на позиции 0.

Шестая строка вызывает метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__getitem__()</span> у списка <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">stuff</span> с параметром 0.


````python
print (stuff.__getitem__(0))
````

Седьмая строка — ещё более многословный способ получения элемента на нулевой позиции в списке.


````python
print (list.__getitem__(stuff,0))
````

В этом коде вызывается метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__getitem__</span> из класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">list</span> и **передаются** сам список и индекс нужного элемента в качестве параметров.

Последние три строки программы эквивалентны, но удобнее просто использовать синтаксис квадратных скобок для обращения к элементу по позиции.

Мы можем посмотреть возможности объекта, взглянув на вывод функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dir()</span>:


````python
stuff = list()
dir(stuff)
````

Остальная часть этой главы определит все приведённые выше термины, поэтому после её прочтения полезно вернуться и перечитать описанные выше абзацы, чтобы проверить своё понимание.

````python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl

# Игнорируем ошибки SSL/TLS сертификата
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
    
url = input('Enter - ')
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')

# Извлекаем все теги привязки
tags = soup('a')
for tag in tags:
    print(tag.get('href', None))
````

Мы считываем URL в строку и передаём её <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">urllib</span> для получения данных из сети. Полученную от <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">urllib</span> строку мы передаём BeautifulSoup для разбора. BeautifulSoup использует объект <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">html.parser</span> и возвращает объект. Затем вызывается метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">tags()</span> для получения словаря объектов-тегов. Проходя в цикле по тегам, мы вызываем метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">get()</span>, чтобы вывести значение атрибута <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">href</span>.

### **Наш первый объект в Python**

На базовом уровне объект — это просто фрагмент кода вместе со структурами данных, меньший, чем целая программа. Определение функции позволяет сохранить фрагмент кода, дать ему имя и затем вызывать этот код по имени функции.

Объект может содержать несколько функций (которые мы называем **методами**), а также данные, которые используют эти функции. Данные, являющиеся частью объекта, называются **атрибутами**.

Для описания данных и кода, составляющих каждый из объектов, мы используем ключевое слово <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">class</span>. Ключевое слово class включает имя класса и начинает отступенный блок кода, в котором содержатся атрибуты (данные) и методы (код).


````python
class PartyAnimal:

   def __init__(self):
     self.x = 0

   def party(self) :
     self.x = self.x + 1
     print("So far",self.x)

an = PartyAnimal()
an.party()
an.party()
an.party()
````

Каждый метод выглядит как функция: начинается с ключевого слова <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">def</span> и состоит из отступенного блока кода.

Первый метод — специально именованный метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__init__()</span>. Он вызывается для выполнения начальной настройки данных, которые мы хотим хранить в объекте. В данном классе мы выделяем атрибут <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> с помощью точечной нотации и инициализируем его нулём.


````python
self.x = 0
````

Второй метод называется <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span>. У всех методов есть специальный первый параметр, который по соглашению называется <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">self</span>. Первый параметр даёт нам доступ к экземпляру объекта, чтобы устанавливать атрибуты и вызывать методы через точечную нотацию.

Как ключевое слово <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">def</span> не приводит к выполнению кода функции, так и ключевое слово <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">class</span> не создаёт объект. <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">class</span> определяет шаблон, указывающий, какие данные и код будут содержаться в каждом объекте типа <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span>. Класс похож на форму для печенья, а объекты, созданные с помощью класса, — это сами печенья. Глазурь мы наносим не на форму, а на готовые печенья, причём на каждое можно нанести разную.

![Класс и два объекта](./PYE4-14-Objects_files/cookie_cutter_flickr_Didriks.png)

*Класс и два объекта*

Если продолжить разбор программы, мы увидим первую исполняемую строку кода:


````python
an = PartyAnimal()
````

Здесь мы указываем Python сконструировать (то есть создать) **объект** или **экземпляр** класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span>. Это выглядит как вызов функции с именем класса. Python конструирует объект с нужными данными и методами и возвращает объект, который затем присваивается переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">an</span>. В некотором смысле это похоже на следующую строку, которую мы используем уже давно:


````python
counts = dict()
````

Здесь мы указываем Python сконструировать объект по шаблону <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dict</span> (уже имеющемуся в Python), вернуть экземпляр словаря и присвоить его переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">counts</span>.

Когда класс <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span> используется для конструирования объекта, переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">an</span> указывает на созданный объект. Мы используем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">an</span> для доступа к коду и данным этого конкретного экземпляра класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span>.

Каждый объект/экземпляр PartyAnimal содержит переменную <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> и метод/функцию <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span>. Вызываем метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span> такой строкой:


````python
an.party()
````

Когда метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span> вызывается, первый параметр (по соглашению <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">self</span>) указывает на конкретный экземпляр объекта PartyAnimal, из которого вызван <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span>. Внутри метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span> мы видим строку:


````python
self.x = self.x + 1
````

Такой синтаксис с точкой означает «x внутри self». При каждом вызове <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party()</span> внутреннее значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> увеличивается на 1 и выводится на экран.

Следующая строка — ещё один способ вызова метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span> у объекта <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">an</span>:


````python
PartyAnimal.party(an)
````

В этом варианте мы обращаемся к коду прямо внутри класса и явно передаём указатель на объект <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">an</span> в качестве первого параметра (то есть <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">self</span> внутри метода). Запись <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">an.party()</span> можно рассматривать как сокращение для такой строки.

При выполнении программа производит следующий вывод:


### **Классы как типы**

Как мы видели, все переменные в Python имеют тип. С помощью встроенной функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dir</span> можно изучить возможности переменной. Функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">type</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dir</span> применимы и к классам, которые мы создаём сами.


````python
class PartyAnimal:

   def __init__(self):
     self.x = 0

   def party(self) :
     self.x = self.x + 1
     print("So far",self.x)

an = PartyAnimal()
print ("Type", type(an))
print ("Dir ", dir(an))
print ("Type", type(an.x))
print ("Type", type(an.party))
````

Из вывода видно, что с помощью ключевого слова <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">class</span> мы создали новый тип. В выводе <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dir</span> видны и целочисленный атрибут <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span>, и метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span>, доступные в объекте.


### **Жизненный цикл объекта**

В предыдущих примерах мы определяем класс (шаблон), используем этот класс для создания экземпляра (объекта), а затем используем этот экземпляр. Когда программа завершается, все переменные отбрасываются. Обычно мы мало задумываемся о создании и уничтожении переменных, однако по мере усложнения объектов нередко необходимо выполнять какие-либо действия при создании объекта и при его уничтожении.

Если мы хотим, чтобы объект был «в курсе» этих моментов, к нему добавляются специально именованные методы:


````python
class PartyAnimal:

   def __init__(self):
     self.x = 0
     print('I am constructed')

   def party(self) :
     self.x = self.x + 1
     print('So far',self.x)

   def __del__(self):
     print('I am destructed', self.x)

an = PartyAnimal()
an.party()
an.party()
an = 42
print('an contains',an)
````

При конструировании объекта Python вызывает наш метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__init__</span>, давая возможность задать значения по умолчанию. Когда Python доходит до строки:

```python
an = 42
```

он фактически «выбрасывает» наш объект, чтобы переиспользовать переменную <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">an</span> для значения <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">42</span>. Именно в момент уничтожения объекта вызывается деструктор (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__del__</span>). Мы не можем предотвратить уничтожение переменной, но можем выполнить необходимую очистку непосредственно перед тем, как объект перестанет существовать.

При разработке объектов очень часто добавляют конструктор для задания начальных значений. Деструктор требуется значительно реже.


### **Несколько экземпляров**

До сих пор мы определяли класс, создавали один объект, использовали его и затем отбрасывали. Однако настоящая сила объектно-ориентированного программирования проявляется тогда, когда мы создаём **несколько экземпляров** нашего класса.

Когда мы создаём несколько объектов из одного класса, возможно, мы захотим задать для каждого свои начальные значения. Можно передавать данные в конструктор, чтобы каждому объекту назначить разные начальные значения:


````python
class PartyAnimal:

   def __init__(self, nam):
     self.x = 0
     self.name = nam
     print(self.name,'constructed')

   def party(self) :
     self.x = self.x + 1
     print(self.name,'party count',self.x)

s = PartyAnimal('Sally')
s.party()
j = PartyAnimal('Jim')

j.party()
s.party()
````

У конструктора есть как параметр <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">self</span>, указывающий на экземпляр объекта, так и дополнительные параметры, которые передаются в конструктор при создании объекта:

```python
s = PartyAnimal('Sally')
```

Внутри конструктора вторая строка копирует переданный параметр (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">nam</span>) в атрибут <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">name</span> внутри экземпляра объекта.

Вывод программы показывает, что каждый из объектов (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">s</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">j</span>) содержит собственные независимые копии <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">name</span>:

```
Sally constructed
Sally party count 1
Jim constructed
Jim party count 1
Sally party count 2
```


### **Наследование**

Ещё одна мощная возможность объектно-ориентированного программирования — создание нового класса путём расширения существующего. При расширении класса исходный класс называют **родительским классом**, а новый — **дочерним классом**.

В нашем примере мы переносим класс <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span> в отдельный файл. После этого можно «импортировать» класс <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span> в новом файле и расширить его так:


````python
class CricketFan(PartyAnimal):

   def __init__(self, nam) :
       super().__init__(nam)
       self.points = 0

   def six(self):
      self.points = self.points + 6
      self.party()
      print(self.name,"points",self.points)

s = PartyAnimal("Sally")
s.party()
j = CricketFan("Jim")
j.party()
j.six()
print(dir(j))
````

Когда мы определяем класс <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">CricketFan</span>, мы указываем, что расширяем класс <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span>. Это означает, что все переменные (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span>) и методы (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span>) класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span> **наследуются** классом <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">CricketFan</span>. Например, внутри метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">six</span> класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">CricketFan</span> мы вызываем метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">party</span> из класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span>.

В методе <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__init__()</span> класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">CricketFan</span> мы используем специальный синтаксис, чтобы также вызвать метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__init__()</span> класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span>. Таким образом выполняется вся настройка, нужная классу <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span>, в дополнение к настройке, необходимой расширениям <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">CricketFan</span>.


````python
def __init__(self, nam) :
       super().__init__(nam)
       self.points = 0
````

Синтаксис <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">super()</span> говорит Python вызвать метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__init__</span> класса, который мы расширяем. <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span> — это супер- (или родительский) класс, а <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">CricketFan</span> — под- (или дочерний) класс.

При выполнении программы создаются <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">s</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">j</span> — независимые экземпляры классов <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">PartyAnimal</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">CricketFan</span>. Объект <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">j</span> обладает дополнительными возможностями по сравнению с объектом <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">s</span>.

В выводе <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dir</span> для объекта <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">j</span> (экземпляр класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">CricketFan</span>) мы видим атрибуты и методы родительского класса, а также атрибуты и методы, добавленные при расширении до класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">CricketFan</span>.


### **Заключение**

Это очень краткое введение в объектно-ориентированное программирование, сосредоточенное главным образом на терминологии и синтаксисе определения и использования объектов. Быстро вернёмся к коду, который мы рассматривали в начале главы. Теперь вы должны полностью понимать, что в нём происходит.


````python
stuff = list()
stuff.append('python')
stuff.append('chuck')
stuff.sort()
print (stuff[0])
print (stuff.__getitem__(0))
print (list.__getitem__(stuff,0))
````

Первая строка конструирует объект <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">list</span>. При создании объекта <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">list</span> Python вызывает **метод-конструктор** (с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__init__</span>) для настройки внутренних атрибутов данных, в которых будет храниться содержимое списка. Мы не передали конструктору никаких параметров. После завершения работы конструктора переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">stuff</span> указывает на возвращённый экземпляр класса <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">list</span>.

Вторая и третья строки вызывают метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">append</span> с одним параметром, чтобы добавить новый элемент в конец списка, обновляя атрибуты внутри <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">stuff</span>. Затем в четвёртой строке вызывается метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sort</span> без параметров, чтобы отсортировать данные внутри объекта <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">stuff</span>.

Далее мы выводим первый элемент списка, используя квадратные скобки — это сокращение вызова метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__getitem__</span> в объекте <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">stuff</span>. Это эквивалентно вызову метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__getitem__</span> в классе <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">list</span> с передачей объекта <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">stuff</span> первым параметром и искомой позиции — вторым.

По завершении программы объект <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">stuff</span> отбрасывается, но предварительно вызывается **деструктор** (с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__del__</span>), чтобы объект мог выполнить необходимую очистку.

Это основы объектно-ориентированного программирования. Существует множество дополнительных деталей о том, как лучше всего применять объектно-ориентированный подход при разработке больших приложений и библиотек, которые выходят за рамки этой главы.


### **Глоссарий**

**Атрибут (attribute)** — одна из именованных единиц данных, составляющих объект.

**Класс (class)** — шаблон, используемый для создания объектов. Определение класса содержит все методы и атрибуты, которые будут иметь объекты этого класса.

**Дочерний класс (child class)** — новый класс, созданный путём расширения существующего. Также называется «подкласс» (subclass).

**Конструктор (constructor)** — необязательный специально именованный метод (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__init__</span>), вызываемый в момент создания объекта для задания начальных значений.

**Деструктор (destructor)** — необязательный специально именованный метод (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">__del__</span>), вызываемый непосредственно перед уничтожением объекта. Деструкторы используются редко.

**Инкапсуляция (encapsulation)** — размещение данных и методов внутри объекта, благодаря чему детали реализации можно скрыть от пользователя объекта.

**Наследование (inheritance)** — создание нового класса (дочернего) с методами и атрибутами, унаследованными от существующего класса (родительского).

**Метод (method)** — функция, содержащаяся внутри класса и его объектов. Иногда называется «операцией». Методы принимают первым параметром <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">self</span>, обычно, через который и получают доступ к самому объекту.

**Объект (object)** — конструкция, созданная на основе класса и содержащая собственные данные (атрибуты) и код (методы).

**Родительский класс (parent class)** — класс, расширяемый для создания нового класса. Родительский класс предоставляет все свои методы и атрибуты новому (дочернему) классу.
