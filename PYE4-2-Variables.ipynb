### **Порядок выполнения операций**

Когда в выражении появляется более одного оператора, порядок вычисления зависит от _правил приоритета_. Для математических операторов Python следует математической конвенции. Акроним _PEMDAS_ полезен для запоминания этих правил:

<ul style="list-style-type: disc; font-weight:bold; margin-left:20px;">
  <li style="font-weight:normal;"><em>P (Parentheses)</em> - Скобки имеют наивысший приоритет и могут использоваться, чтобы заставить выражение вычисляться в нужном вам порядке. Поскольку выражения в скобках вычисляются первыми, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">2 * (3-1)</span> равно 4, а <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">(1+1)**(5-2)</span> равно 8. Скобки также можно использовать, чтобы сделать выражение более читаемым, как в <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">(minute * 100) / 60</span>, даже если это не меняет результата.</li>
  <li style="font-weight:normal;"><em>E (Exponentiation)</em> - Возведение в степень имеет следующий по величине приоритет, поэтому <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">2**1+1</span> равно 3, а не 4, и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">3*1**3</span> равно 3, а не 27.</li>
  <li style="font-weight:normal;"><em>M (Multiplication)</em> и <em>D (Division)</em> - Умножение и деление имеют одинаковый приоритет, который выше, чем у сложения (<em>A - Addition</em>) и вычитания (<em>S - Subtraction</em>), которые также имеют одинаковый приоритет. Поэтому <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">2*3-1</span> равно 5, а не 4, и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">6+4/2</span> равно 8, а не 5.</li>
  <li style="font-weight:normal;">Операторы с одинаковым приоритетом вычисляются слева направо. Поэтому выражение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">5-3-1</span> равно 1, а не 3, потому что сначала выполняется <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">5-3</span>, а затем от результата вычитается 1.</li>
</ul>

Если есть сомнения, всегда используйте скобки в выражениях, чтобы убедиться, что вычисления выполняются в нужном вам порядке.

### **Оператор взятия остатка**

_Оператор взятия остатка_ работает с целыми числами и возвращает остаток от деления первого операнда на второй. В Python оператор взятия остатка обозначается знаком процента (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">%</span>). Синтаксис такой же, как и для других операторов:

````python
quotient = 7 // 3 # Оператор для получения целой части от деления
print(quotient)

remainder = 7 % 3 # Оператор для получения остатка от деления (дробной части)
print(remainder)
````

Таким образом, 7, делённое на 3, даёт 2 с остатком 1.

Оператор взятия остатка оказывается неожиданно полезным. Например, с его помощью можно проверить, делится ли одно число на другое: если <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x % y</span> равно нулю, то <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> делится на <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">y</span>.

Также с его помощью можно получить правую (самую младшую) цифру или цифры числа. Например, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x % 10</span> даёт последнюю цифру числа <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> (в десятичной системе).

Аналогично, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x % 100</span> даёт последние две цифры.

### **Операции со строками**

Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">+</span> работает со строками, но это не сложение в математическом смысле. Вместо этого он выполняет конкатенацию, то есть соединяет строки, связывая их друг с другом. Например:

````python
first = 10
second = 15
print(first+second)

first = '100'
second = '150'
print(first + second) # В контексте сложения строк происходит конкатенация, или сложение строк, как если бы мы их склеили
````

````python
first = 'Test '
second = 3
print(first * second)
````

### **Запрос ввода у пользователя**

Иногда нам нужно получить значение переменной от пользователя через клавиатуру. В Python есть встроенная функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">input</span>, которая получает ввод с клавиатуры. Когда эта функция вызывается, программа останавливается и ждёт, пока пользователь что-нибудь введёт. Когда пользователь нажимает <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">Return</span> или <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">Enter</span>, программа продолжает выполнение, а <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">input</span> возвращает то, что ввёл пользователь, в виде строки.

````python
inp = input()
print(inp)
````

````python
name = input('What is your name?\n') # \n - это строковый оператор перевода на новую строку
print(name)
````

Последовательность <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">\n</span> в конце запроса обозначает <em>перевод строки</em> — это специальный символ, который вызывает переход на новую строку. ### **Комментарии**

По мере того как программы становятся больше и сложнее, их становится труднее читать. Формальные языки плотные, и часто бывает сложно, глядя на фрагмент кода, понять, что он делает или почему.

По этой причине полезно добавлять в программы пояснения на естественном языке о том, что делает программа. Эти пояснения называются <em>комментариями</em>, и в Python они начинаются с символа <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">#</span>:

````python
# Вычисление процента из часа
percentage = (minute * 100) / 60
````

В этом случае комментарий расположен на отдельной строке. Комментарии также можно размещать в конце строки:

````python
# Вычисление процента из часа
percentage = (minute * 100) / 60 # Процент от часа
````

Всё, что находится после символа <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">#</span> до конца строки, игнорируется; это никак не влияет на выполнение программы.

Комментарии наиболее полезны, когда они объясняют неочевидные особенности кода. Можно предположить, что читатель разберётся, что делает код; гораздо полезнее объяснить, <em>почему он это делает</em>.

Этот комментарий дублирует код и бесполезен:

````python
v = 5 # Присваиваем 5 переменной v
````

Этот комментарий содержит полезную информацию, которой нет в коде:

````python
v = 5 # Скорость в метрах в секунду
````

Хорошие имена переменных могут уменьшить необходимость в комментариях, но длинные имена могут усложнить чтение сложных выражений, поэтому здесь существует компромисс.

Интерпретатор Python воспринимает все три программы как абсолютно одинаковые, но люди видят и понимают их совершенно по-разному. Человек быстрее всего поймёт назначение второй программы, потому что программист выбрал имена переменных, отражающие смысл того, какие данные будут храниться в каждой переменной.

Такие удачно подобранные имена переменных называют «мнемоническими именами переменных». Слово mnemonic означает «средство для запоминания». Мы выбираем мнемонические имена переменных, чтобы помнить, зачем вообще создали переменную.

Хотя это звучит отлично, и использовать мнемонические имена действительно хорошая идея, они могут мешать начинающему программисту разбирать и понимать код. Это связано с тем, что новички ещё не запомнили зарезервированные слова (их всего 35), и иногда слишком описательные имена переменных начинают казаться частью языка, а не просто удачно выбранными именами.

Посмотрите на следующий пример кода Python, который перебирает некоторые данные. Циклы мы рассмотрим позже, а пока попробуйте просто понять, что здесь происходит:

````python
words = "Hello, World!"
for word in words:
    print(word)
````

Что здесь происходит? Какие из элементов (for, word, in и т. д.) являются зарезервированными словами, а какие — просто именами переменных? Понимает ли Python на базовом уровне понятие слов? Начинающим программистам сложно отделить части кода, которые должны быть такими же, как в примере, от тех частей, которые являются просто выбором программиста.

Следующий код эквивалентен приведённому выше:

````python
pizza = "Pizza"
for slice in pizza:
    print(slice)
````

Начинающему программисту проще посмотреть на этот код и понять, какие части являются зарезервированными словами Python, а какие — просто именами переменных, выбранными программистом. Довольно очевидно, что Python не имеет фундаментального понимания пиццы и кусков и того факта, что пицца состоит из одного или нескольких кусков.

Но если программа действительно предназначена для чтения данных и поиска слов в этих данных, то pizza и slice — крайне немнемонические имена переменных. Их использование отвлекает от смысла программы.

Спустя довольно короткое время вы выучите наиболее распространённые зарезервированные слова и начнёте сразу замечать их в коде:

````python
for word in words:
    print(word)
````

Части кода, определённые Python (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">in</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">:</span>), выделены жирным, а переменные, выбранные программистом (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">word</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">words</span>), — нет. Многие текстовые редакторы понимают синтаксис Python и подсвечивают зарезервированные слова другим цветом, помогая вам отличать переменные от зарезервированных слов. Со временем вы начнёте читать Python и быстро определять, что является переменной, а что — зарезервированным словом.

### **Отладка**

На этом этапе наиболее вероятная синтаксическая ошибка — использование недопустимого имени переменной, например <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">class</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">yield</span>, которые являются ключевыми словами, или <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">odd~job</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">US$</span>, которые содержат недопустимые символы.

Если вы вставите пробел в имени переменной, Python решит, что это два операнда без оператора:

````python
# bad name = 5
# Не будет работать, из-за пробела между словами, переменные должны быть одним словом
````

При синтаксических ошибках сообщения об ошибке мало помогают. Наиболее распространённое сообщение — <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">SyntaxError: invalid syntax</span>, что не слишком информативно.

Ошибка времени выполнения, которую вы, скорее всего, допустите, — это «использование до определения»; то есть попытка использовать переменную до того, как ей было присвоено значение. Это может произойти, если вы неправильно напишете имя переменной:

````python
# principal = 327.68
# interest = principle * rate

# Не будет работать, из-за неопреденной переменной principle (То что было объявлено отличается от значения в выражении)
````

Имена переменных чувствительны к регистру, поэтому <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">LaTeX</span> — это не то же самое, что <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">latex</span>.

На данном этапе наиболее вероятной причиной семантической ошибки является порядок выполнения операций. Например, чтобы вычислить 1/2π, у вас может возникнуть соблазн написать:

````python
1.0 / 2.0 * pi
# Тут смысл ошибки в порядке операторов, хотели по сути 1/2pi, получили pi/2
````

Но деление выполняется первым, поэтому вы получите π/2, что не одно и то же! Python никак не может понять, что именно вы имели в виду, поэтому в этом случае сообщение об ошибке не появляется; вы просто получаете неверный результат.

### **Глоссарий**

**присваивание** - 
Инструкция, которая присваивает значение переменной.

**конкатенация** - 
Соединение двух операндов друг с другом.

**комментарий** - 
Информация в программе, предназначенная для других программистов (или любого, кто читает исходный код) и не влияющая на выполнение программы.

**вычислять** - 
Упрощать выражение, выполняя операции, чтобы получить одно итоговое значение.

**выражение** - 
Комбинация переменных, операторов и значений, представляющая одно результирующее значение.

**число с плавающей точкой** - 
Тип, представляющий числа с дробной частью.

**целое число** - 
Тип, представляющий целые числа.

**ключевое слово** - 
Зарезервированное слово, используемое компилятором для разбора программы; вы не можете использовать такие ключевые слова, как if, def и while, в качестве имён переменных.

**мнемонический** - 
Средство для запоминания. Мы часто даём переменным мнемонические имена, чтобы помнить, что в них хранится.

**оператор взятия остатка** - 
Оператор, обозначаемый знаком процента (%), который работает с целыми числами и возвращает остаток при делении одного числа на другое.

**операнд** - 
Одно из значений, над которым работает оператор.

**оператор** - 
Специальный символ, обозначающий простое вычисление, такое как сложение, умножение или конкатенация строк.

**правила приоритета** - 
Набор правил, определяющих порядок вычисления выражений, содержащих несколько операторов и операндов.

**инструкция** - 
Фрагмент кода, представляющий команду или действие. До сих пор мы видели инструкции присваивания и инструкцию вывода выражения print.

**строка** - 
Тип, представляющий последовательности символов.

**тип** - 
Категория значений. До сих пор мы рассматривали целые числа (тип int), числа с плавающей точкой (тип float) и строки (тип str).

**значение** - 
Одна из базовых единиц данных, например число или строка, с которой работает программа.

**переменная** - 
Имя, которое ссылается на значение.
