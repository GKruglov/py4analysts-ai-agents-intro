### **Логические выражения**

<em>Логическое выражение</em> — это выражение, которое является либо истинным, либо ложным. Следующие примеры используют оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">==</span>, который сравнивает два операнда и возвращает <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">True</span>, если они равны, и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">False</span> в противном случае:

````python
print(5 == 5, 5 == 6)
````

<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">True</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">False</span> — это специальные значения, которые принадлежат классу <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">bool</span>; они не являются строками:

````python
print(type(True), type(False))
````

Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">==</span> является одним из _операторов сравнения_; другие операторы:

````python
# x != y               x не равен y
# x > y                x больше, чем y
# x < y                x меньше, чем y
# x >= y               x больше или равен y
# x <= y               x меньше или равен y
# x is y               x такой же как y
# x is not y           x не такой же как y
# Разница == от is в том, что первый оператор проверяет равны ли объекты по содержанию, а второй задает вопрос "это один и тот же объект в памяти?"
````

Хотя эти операции, вероятно, вам знакомы, символы Python отличаются от математических символов для тех же операций. Распространённая ошибка — использовать один знак равенства (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">=</span>) вместо двойного знака равенства (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">==</span>). Помните, что <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">=</span> — это оператор присваивания, а <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">==</span> — оператор сравнения. Не существует таких записей, как <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">=<</span> или <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">=></span>.

### **Логические операторы**

Существует три _логических оператора_: <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">and</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">or</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">not</span>. Семантика (значение) этих операторов похожа на их значение в английском языке. Например, 

<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x > 0 and x < 10</span> истинно только если <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> больше 0 _и_ меньше 10.

<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">n%2 == 0 or n%3 == 0</span> истинно, если выполняется _хотя бы одно_ из условий, то есть если число делится на 2 _или_ на 3.

Наконец, оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">not</span> отрицaет логическое выражение, поэтому <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">not (x > y)</span>истинно, если <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x > y</span> ложно.

````python
x = 1
y = 2
x > y
````

````python
not (x > y)
````

Строго говоря, операнды логических операторов должны быть логическими выражениями, но Python не очень строг. Любое ненулевое число интерпретируется как «истина».

````python
17 and True
````

Такая гибкость может быть полезна в некоторых ситуациях, но в этом есть нюансы, которые могут запутать. Возможно, стоит избегать этого, пока вы не будете уверены, что понимаете, что делаете.

### **Условное выполнение** ###

Чтобы писать полезные программы, нам почти всегда нужна возможность проверять условия и изменять поведение программы в соответствии с ними. *Условные операторы* дают нам эту возможность. Самая простая форма — оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span>:

````python
if x > 0 :
    print('x is positive')
````

Логическое выражение после оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span> называется _условием_. Мы заканчиваем строку с <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span> двоеточием (:) и строки после if должны быть с отступом.

Логика if

Если логическое условие истинно, то выполняется блок с отступом. Если логическое условие ложно, блок с отступом пропускается.

Операторы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span> имеют ту же структуру, что и определения функций или циклы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>. Оператор состоит из заголовка, который заканчивается двоеточием (:), и блока с отступом. Такие операторы называются *составными*, потому что они занимают более одной строки.


````python
if x > y:
    print(x)
    print(y)
````

Нет ограничения на количество операторов в теле, но тело должно содержать хотя бы один оператор. Иногда полезно иметь тело без операторов (обычно как заполнитель для кода, который ещё не написан). В этом случае можно использовать оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">pass</span>, который проходит проверку интерпретатора Python, но ничего не делает.

````python
if x < 0 :
    pass   # Нужно обработать отрицательные значения, пока ничего не делает
````

Если вы вводите оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span> в интерпретаторе Python, приглашение изменится с трёх символов больше (>>>) на три точки (…) чтобы показать, что вы находитесь внутри блока операторов:

````python
x = 3
if x < 10:
    print('Small')
````

При использовании интерпретатора Python необходимо оставить пустую строку в конце блока, иначе Python вернёт ошибку:

````python
x = 3
if x < 10:
    print('Small')
 print('Done')
````

Пустая строка в конце блока операторов не обязательна при написании и выполнении скрипта, но может улучшить читаемость кода.

### **Альтернативное выполнение**

Вторая форма оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span> — это *альтернативное выполнение*, при котором есть две возможности, и условие определяет, какая из них будет выполнена. Синтаксис выглядит так:

````python
if x % 2 == 0:
    print('x is even')
else:
    print('x is odd')
````

Если остаток от деления <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> на 2 равен 0, значит, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> чётное, и программа выводит соответствующее сообщение. Если условие ложно, выполняется второй набор операторов.

Логика If-Then-Else

Поскольку условие должно быть либо истинным, либо ложным, выполняется ровно одна из альтернатив. Эти альтернативы называются *ветвями*, потому что они представляют собой разветвления в потоке выполнения.

### **Цепочка условий**

Иногда возможностей больше двух, и нам нужно больше двух ветвей. Один из способов выразить такую логику — это *цепочка условий*:

````python
if x < y:
    print('x is less than y')
elif x > y:
    print('x is greater than y')
else:
    print('x and y are equal')
````

<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">elif</span> — сокращение от «else if». Опять же, выполняется ровно одна ветвь.

Логика If-Then-ElseIf

Нет ограничения на количество операторов <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">elif</span>. Если есть блок <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">else</span>, он должен быть в конце, но он необязателен.

````python
choice = 'b'
if choice == 'a':
    print('Bad guess')
elif choice == 'b':
    print('Good guess')
elif choice == 'c':
    print('Close, but not correct')
````

Каждое условие проверяется по порядку. Если первое ложно, проверяется следующее и так далее. Если одно из условий истинно, выполняется соответствующая ветвь, и оператор заканчивается. Даже если несколько условий истинны, выполняется только первая истинная ветвь.

### **Вложенные условия**

Один условный оператор может быть вложен в другой. Мы могли бы записать пример с тремя ветвями так:

````python
if x == y:
    print('x and y are equal')
else:
    if x < y:
        print('x is less than y')
    else:
        print('x is greater than y')
````

Внешний условный оператор содержит две ветви. Первая ветвь содержит простой оператор. Вторая ветвь содержит ещё один оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span>, который имеет две свои ветви. Обе эти ветви — простые операторы, хотя они тоже могли быть условными операторами.

Вложенные операторы If

Хотя отступы делают структуру очевидной, *вложенные условия* быстро становятся трудно читаемыми. В общем, лучше избегать их, когда это возможно.

Логические операторы часто позволяют упростить вложенные условия. Например, следующий код можно переписать с использованием одного условия:

````python
if 0 < x:
    if x < 10:
        print('x is a positive single-digit number.')
````

Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span> выполняется только если выполняются оба условия. Такой же эффект можно получить с помощью оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">and</span>:

````python
if 0 < x and x < 10:
    print('x is a positive single-digit number.')
````

### **Обработка исключений с помощью try и except**

Ранее мы видели фрагмент кода, где использовались функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">input</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">int</span> для чтения и преобразования введённого пользователем целого числа. Мы также видели, насколько это может быть опасно:

````python
prompt = "What is the air velocity of an unladen swallow?\n"
speed = input(prompt)
int(speed)
````

Когда мы выполняем эти инструкции в интерпретаторе Python, мы получаем новый приглашение от интерпретатора, думаем «ой» и продолжаем с следующей командой.

Однако если разместить этот код в скрипте Python и произойдёт ошибка, скрипт сразу остановится с трассировкой. Следующая инструкция не будет выполнена.

Пример программы для преобразования температуры из Фаренгейта в Цельсий:

````python
inp = input('Enter Fahrenheit Temperature: ')
fahr = float(inp)
cel = (fahr - 32.0) * 5.0 / 9.0
print(cel)
````

Если выполнить этот код и ввести некорректное значение, программа завершится с ошибкой:

````python
inp = input('Enter Fahrenheit Temperature: ')
fahr = float(inp)
cel = (fahr - 32.0) * 5.0 / 9.0
print(cel)
````

В Python есть структура условного выполнения для обработки ожидаемых и неожиданных ошибок — это try / except. Цель <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">try</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">except</span> в том, что вы знаете, что какая-то последовательность инструкций может вызвать проблему, и хотите добавить инструкции, которые будут выполнены в случае ошибки. Эти дополнительные инструкции (блок except) игнорируются, если ошибки нет.

Можно рассматривать <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">try</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">except</span> как «страховку» для последовательности операторов.

Мы можем переписать наш конвертер температуры следующим образом:

````python
inp = input('Enter Fahrenheit Temperature:')
try:
    fahr = float(inp)
    cel = (fahr - 32.0) * 5.0 / 9.0
    print(cel)
except:
    print('Please enter a number')
````

Python начинает выполнение операторов в блоке <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">try</span>. Если всё идёт хорошо, блок <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">except</span> пропускается, и выполнение продолжается. Если в блоке <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">try</span> возникает исключение, Python выходит из блока try и выполняет последовательность операторов в блоке <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">except</span>.

````python
inp = input('Enter Fahrenheit Temperature:')
try:
    fahr = float(inp)
    cel = (fahr - 32.0) * 5.0 / 9.0
    print(cel)
except:
    print('Please enter a number')
````

Обработка исключения с помощью оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">try</span> называется *перехватом исключения*. В этом примере блок <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">except</span> выводит сообщение об ошибке. В общем случае перехват исключения даёт возможность исправить проблему, повторить попытку или, по крайней мере, завершить программу корректно.

### **Короткая логическая оценка выражений**

Когда Python обрабатывает логическое выражение, например <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x >= 2 and (x/y) > 2</span> он оценивает его слева направо. Из-за определения оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">and</span>, если <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> меньше 2, выражение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x >= 2</span> **ложно**, и всё выражение становится **ложным**, независимо от того, будет ли <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">(x/y) > 2</span> **истинным** или **ложным**.

Когда Python обнаруживает, что дальнейшая оценка логического выражения не даст никакой пользы, он прекращает вычисления и не выполняет оставшиеся части выражения. Когда оценка логического выражения останавливается, потому что общий результат уже известен, это называется *короткой логической оценкой (short-circuiting)*.

Хотя это может показаться незначительным моментом, поведение короткой оценки позволяет использовать приём, называемый *шаблоном «стража» (guardian pattern)*. Рассмотрим следующий код в интерпретаторе Python:

````python
x = 6
y = 2
x >= 2 and (x / y) > 2
````

````python
x = 1
y = 0
x >= 2 and (x / y) > 2
````

````python
x = 6
y = 0
x >= 2 and (x / y) > 2
````

Третье вычисление завершилось с ошибкой, потому что Python пытался вычислить <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">(x/y) </span>и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">y</span> было равно нулю, что вызывает ошибку времени выполнения. Но первые два примера *не* вызвали ошибку: в первом <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">y</span> было ненулевым, а во втором первая часть выражения <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x >= 2</span> оказалась **ложной**, поэтому <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">(x/y)</span> даже не вычислялось из-за правила *короткой логической оценки*.

Мы можем построить логическое выражение так, чтобы разместить *проверку-«стража»* перед вычислением, которое может вызвать ошибку:

````python
x = 1
y = 0
x >= 2 and y != 0 and (x / y) > 2
````

````python
x = 6
y = 0
x >= 2 and y != 0 and (x / y) > 2
````

````python
x >= 2 and (x/y) > 2 and y != 0
````

В первом выражении <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x >= 2</span> **ложно**, поэтому оценка останавливается на <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">and</span>. Во втором выражении <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x >= 2</span> **истинно**, но <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">y != 0</span> **ложно**, поэтому мы никогда не доходим до <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">(x/y)</span>.

В третьем выражении <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">y != 0</span> стоит после вычисления <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">(x/y)</span>, поэтому выражение завершается ошибкой.

Во втором выражении можно сказать, что <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">y != 0</span> действует как *«страж»*, который гарантирует, что <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">(x/y)</span> выполняется только если <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">y</span> ненулевое.

### **Отладка**

Трассировка (traceback), которую Python выводит при возникновении ошибки, содержит много информации, но она может быть перегружающей. Обычно наиболее полезные части:
<ul style="list-style-type: disc; font-weight:bold; margin-left:20px;">
  <li style="font-weight:normal;">Какого типа была ошибка, и</li>
  <li style="font-weight:normal;">Где она произошла.</li>
  
</ul>

Синтаксические ошибки обычно легко найти, но есть несколько подвохов. Ошибки с пробелами могут быть коварными, потому что пробелы и табуляции невидимы, а мы привыкли их игнорировать.

````python
x = 5
 y = 6
````

В этом примере проблема в том, что вторая строка сдвинута на один пробел. Но сообщение об ошибке указывает на <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">y</span>, что может вводить в заблуждение. В общем, сообщения об ошибках показывают, где была обнаружена проблема, но фактическая ошибка могла быть ранее в коде, иногда на предыдущей строке.

В целом, сообщения об ошибках показывают место обнаружения проблемы, но это не всегда место её возникновения.

### **Словарь терминов**

**body** — последовательность операторов внутри составного оператора.

**boolean expression** — выражение, значение которого либо <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">True</span>, либо <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">False</span>.

**branch** — одна из альтернативных последовательностей операторов в условном операторе.

**chained conditional** — условный оператор с серией альтернативных ветвей.

**comparison operator** — один из операторов, который сравнивает операнды: <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">==</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">!=</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">></span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;"><</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">>=</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;"><=</span>.

**conditional statement** — оператор, который управляет потоком выполнения в зависимости от условия.

**condition** — логическое выражение в условном операторе, определяющее, какая ветвь выполняется.

**compound statement** — оператор, состоящий из заголовка и тела. Заголовок заканчивается двоеточием (:). Тело имеет отступ относительно заголовка.

**guardian pattern** — когда логическое выражение строится с дополнительными проверками, чтобы использовать поведение короткой логической оценки.

**logical operator** — один из операторов, который комбинирует логические выражения: <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">and</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">or</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">not</span>.

**nested conditional** — условный оператор, который находится внутри одной из ветвей другого условного оператора.

**traceback** — список выполняющихся функций, выводимый при возникновении исключения.

**short circuit** — когда Python прерывает вычисление логического выражения, потому что уже известен итоговый результат без необходимости вычислять оставшуюся часть выражения.

### **Упражнения**

**Упражнение 1**: Перепишите расчёт зарплаты так, чтобы сотрудник получал в 1,5 раза больше за часы сверх 40.

``` Python
Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```

````python
hours = int(input("Enter hours"))
rate = float(input("Enter rate"))
print(f'Enter Hours {hours}')
print(f'Enter rate {rate}')
if hours > 40:
    pay = 40 * rate + (hours - 40 ) * (rate * 1.5)
    print(pay)
else:
    print(f"Pay: {hours * rate}")
````

**Упражнение 2**: Перепишите программу расчёта зарплаты с использованием <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">try</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">except</span>, чтобы программа корректно обрабатывала нечисловой ввод, выводя сообщение об ошибке и завершение программы. Пример двух запусков:

``` Python
Enter Hours: 20
Enter Rate: nine
Error, please enter numeric input
```

``` Python
Enter Hours: forty
Error, please enter numeric input
```

````python
try:
    hours = int(input("Enter Hours: "))
    rate = float(input("Enter Rate: "))

    if hours > 40:
        pay = 40 * rate + (hours - 40) * (rate * 1.5)
    else:
        pay = hours * rate
        
    print(f"Pay: {pay}")

except ValueError:
    print("Error, please enter numeric input.")
````

Запускайте программу несколько раз, как показано выше, чтобы проверить работу с разными вариантами ввода:

````python
raw_score = input("Enter a score")

try:
    score = float(raw_score)
    
    if not (0.0 <= score <= 1.0):
        print("Score must be between 0.0 and 1.0!")
    else:        
        if score >= 0.9:
            grade = 'A'
        elif score >= 0.8:
            grade = 'B'
        elif score >= 0.7:
            grade = 'C'
        elif score >= 0.6:
            grade = 'D'
        else:
            grade = 'F'
        
        print(f"Grade: {grade}")

except ValueError:
    print("Bad score")
````
