## **Функции**
### **Вызов функции**

В контексте программирования _функция_ — это именованная последовательность инструкций (операторов), которая выполняет некоторое вычисление. При определении функции вы задаёте её имя и описываете последовательность инструкций, которые она содержит. Позже вы можете "вызвать" функцию по её имени.

Мы уже видели один пример _вызова функции_:


```python
print(type(32))
```

Имя этой функции — <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">type</span>. Выражение в круглых скобках называется _аргументом_ функции. Аргумент — это значение или переменная, которые передаются в функцию в качестве входных данных.

Для функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">type</span> результатом является тип переданного аргумента.

Принято говорить, что функция "принимает" (takes) аргумент и "возвращает" (returns) результат. Возвращаемый результат называется _возвращаемым значением_ (return value).

### **Встроенные функции**

Python предоставляет ряд важных встроенных функций, которые можно использовать без необходимости самостоятельно определять их. Создатели Python реализовали набор функций для решения типовых задач и включили их в язык, чтобы ими можно было пользоваться сразу.

Функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">max</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">min</span> возвращают соответственно наибольшее и наименьшее значение в списке:

```python
max('Hello world')
```

```python
min('Hello world')
```

Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">max</span> сообщает нам «наибольший символ» в строке (в данном случае это оказывается буква «w»), а функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">min</span> показывает наименьший символ (в данном случае — пробел).

Ещё одна очень распространённая встроенная функция — <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len</span>. Она возвращает количество элементов в переданном ей аргументе. Если аргументом функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len</span> является строка, то она возвращает количество символов в этой строке.

```python
len('Hello world')
```

Эти функции не ограничиваются работой только со строками. Они могут применяться к любому набору значений, как мы увидим в следующих главах.

Имена встроенных функций следует рассматривать как зарезервированные слова, то есть не использовать их в качестве имён переменных (например, не стоит называть переменную "max").

Python также предоставляет встроенные функции, которые преобразуют значения из одного типа в другой. Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">int</span> принимает любое значение и преобразует его в целое число, если это возможно, в противном случае возникает ошибка.

Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">float</span> преобразует целые числа и строки в числа с плавающей запятой:

### **Математические функции**

В Python есть модуль <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">math</span>, который предоставляет большинство привычных математических функций. Прежде чем использовать этот модуль, его необходимо импортировать:

```python
import math
```

Этот оператор создаёт _объект модуля_ с именем math. Если вывести этот объект модуля на экран, можно получить некоторую информацию о нём:

```python
print(math)
```

Объект модуля содержит функции и переменные, определённые внутри этого модуля. Чтобы обратиться к одной из функций, необходимо указать имя модуля и имя функции, разделив их точкой (dot, или period). Такой способ записи называется _точечной нотацией_ (dot notation).

В первом примере вычисляется десятичный логарифм (по основанию 10) отношения сигнал/шум. Модуль math также предоставляет функцию <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">log</span>, которая вычисляет логарифм по основанию e (натуральный логарифм).

Во втором примере находится синус угла, заданного в <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">радианах</span>. Имя переменной подсказывает, что функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sin</span> и другие тригонометрические функции (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">cos</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">tan</span> и т. д.) принимают аргументы в радианах. Если вы знакомы с тригонометрией, можно проверить предыдущий результат, сравнив его с квадратным корнем из двух, делённым на два:

### **Случайные числа**

При одинаковых входных данных большинство компьютерных программ каждый раз выдают одинаковые результаты, поэтому их называют _детерминированными_. Детерминированность обычно является положительным свойством, поскольку мы ожидаем, что одно и то же вычисление приведёт к одному и тому же результату. Однако для некоторых задач требуется непредсказуемость. Очевидный пример — игры, но существуют и другие случаи.

Создать по-настоящему недетерминированную программу на практике непросто, однако есть способы сделать её поведение внешне _недетерминированным_. Один из них — использование _алгоритмов_, генерирующих _псевдослучайные_ числа. Псевдослучайные числа не являются по-настоящему случайными, поскольку они получаются в результате детерминированных вычислений, но по их значениям практически невозможно отличить их от истинно случайных.

Модуль <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">random</span> предоставляет функции для генерации псевдослучайных чисел (далее они будут называться просто "случайными").

Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">random</span> возвращает случайное число с плавающей запятой в диапазоне от 0.0 до 1.0 (включая 0.0, но не включая 1.0). Каждый вызов <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">random</span> возвращает следующее число из длинной последовательности. Чтобы увидеть пример, выполните следующий цикл:

```python
import random

for i in range(10):
    x = random.random()
    print(x)
```

Эта программа выводит следующий список из 10 случайных чисел в диапазоне от 0.0 до 1.0 (не включая 1.0).

Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">random</span> — лишь одна из многих функций для работы со случайными числами. Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">randint</span> принимает параметры <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">low</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">high</span> и возвращает целое число в диапазоне от <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">low</span> до <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">high</span> включительно.

```python
random.randint(5, 10)
```

Чтобы случайным образом выбрать элемент из последовательности, можно использовать функцию <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">choice</span>:

```python
t = [1, 2, 3]
random.choice(t)
```

Модуль <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">random</span> также предоставляет функции для генерации случайных значений из непрерывных распределений, включая нормальное (гауссово), экспоненциальное, гамма-распределение и некоторые другие.

### **Добавление новых функций**

До этого момента мы использовали только функции, которые уже входят в состав Python, однако в языке также можно создавать собственные функции. _Определение функции_ задаёт имя новой функции и последовательность инструкций, которые будут выполняться при её вызове. После определения функцию можно многократно использовать в разных частях программы.

Ниже приведён пример:

```python
def print_lyrics():
    print("I'm a lumberjack, and I'm okay.")
    print('I sleep all night and I work all day.')
```

<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">def</span> — это ключевое слово, указывающее на определение функции. Имя функции в данном примере — <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print_lyrics</span>. Правила именования функций такие же, как и для переменных: допускаются буквы, цифры и некоторые знаки пунктуации, однако имя не может начинаться с цифры. Нельзя использовать ключевые слова в качестве имени функции, а также следует избегать ситуации, когда переменная и функция имеют одно и то же имя.

Пустые круглые скобки после имени означают, что функция не принимает аргументов. Позже будут рассмотрены функции, которые принимают аргументы в качестве входных данных.

Первая строка определения функции называется _заголовком_ (_header_), остальная часть — _телом_ функции (_body_). Заголовок должен заканчиваться двоеточием, а тело — быть записано с отступом. По принятому соглашению используется отступ в четыре пробела. Тело функции может содержать любое количество инструкций.

Если вводить определение функции в интерактивном режиме, интерпретатор выводит многоточие (...), показывая, что определение ещё не завершено:


Чтобы завершить определение функции в интерактивном режиме, нужно ввести пустую строку (в файле-скрипте это не требуется).

Определение функции создаёт переменную с тем же именем, что и функция.


```python
print(print_lyrics)
print(type(print_lyrics))
```

Значением <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print_lyrics</span> является _объект функции_, имеющий тип _function_.

Синтаксис вызова новой функции такой же, как и у встроенных функций:


```python
print_lyrics()
```


После определения функции её можно использовать внутри другой функции. Например, чтобы повторить предыдущий припев, можно написать функцию с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">repeat_lyrics</span>:


```python
def repeat_lyrics():
    print_lyrics()
    print_lyrics()
```

А затем вызвать <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">repeat_lyrics</span>:


```python
repeat_lyrics()
```

Однако, на самом деле песня звучит не совсем так.


### **Определения и вызовы**

Если объединить фрагменты кода из предыдущего раздела, полная программа будет выглядеть следующим образом:


```python
def print_lyrics():
    print("I'm a lumberjack, and I'm okay.")
    print('I sleep all night and I work all day.')


def repeat_lyrics():
    print_lyrics()
    print_lyrics()

repeat_lyrics()
```

В этой программе определены две функции: <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print_lyrics</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">repeat_lyrics</span>. Определения функций выполняются так же, как и другие инструкции, однако их результат — создание объектов функций. Инструкции внутри функции не выполняются до тех пор, пока функция не будет вызвана, и само определение функции не выводит никаких данных.

Как можно ожидать, прежде чем выполнить функцию, её необходимо создать. Иными словами, определение функции должно быть выполнено до первого её вызова.

### **Поток выполнения**

Чтобы гарантировать, что функция определена до её первого использования, необходимо понимать порядок выполнения инструкций, который называется _потоком выполнения_ (flow of execution).

Выполнение программы всегда начинается с первой инструкции. Инструкции выполняются последовательно, по одной, сверху вниз.

_Определения_ функций не изменяют общий поток выполнения программы, однако следует помнить, что инструкции внутри функции не выполняются до её вызова.

Вызов функции можно представить как отклонение от основного потока выполнения. Вместо перехода к следующей инструкции программа переходит к телу функции, выполняет все содержащиеся в ней инструкции, а затем возвращается к тому месту, где была прервана.

На первый взгляд это просто, но стоит учитывать, что одна функция может вызывать другую. Во время выполнения одной функции программа может перейти к выполнению другой. В свою очередь, эта функция может вызвать ещё одну.

К счастью, Python корректно отслеживает текущее состояние выполнения, поэтому после завершения каждой функции программа продолжает работу с того места в вызывающей функции, где она остановилась. Когда достигается конец программы, её выполнение завершается.

Вывод из этого следующий: при чтении программы не всегда достаточно двигаться строго сверху вниз. Иногда более разумно следовать потоку выполнения.

### **Параметры и аргументы**

Некоторые из встроенных функций, которые мы рассматривали, требуют аргументы. Например, при вызове <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">math.sin</span> вы передаёте число в качестве аргумента. Некоторые функции принимают более одного аргумента: <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">math.pow</span> принимает два — основание и показатель степени.

Внутри функции переданные аргументы присваиваются переменным, которые называются параметрами. Ниже приведён пример пользовательской функции, принимающей аргумент:


```python
def print_twice(bruce):
    print(bruce)
    print(bruce)
```

Эта функция присваивает переданный аргумент параметру с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">bruce</span>. Когда функция вызывается, она дважды выводит значение этого параметра (каким бы оно ни было).

Эта функция работает с любым значением, которое можно вывести на экран.


```python
print_twice('Spam')
print_twice(17)
```

```python
import math
print_twice(math.pi)
```

Те же правила композиции, которые применяются к встроенным функциям, действуют и для пользовательских функций, поэтому в качестве аргумента для <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print_twice</span> можно использовать выражение любого типа.


```python
print_twice('Spam '*4)
print_twice(math.cos(math.pi))
```

Аргумент вычисляется до вызова функции, поэтому в приведённых примерах выражения <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">'Spam ' * 4</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">math.cos</span>(<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">math.pi</span>) вычисляются только один раз.

В качестве аргумента также можно использовать переменную:


```python
michael = 'Eric, the half a bee.'
print_twice(michael)
```

Имя переменной, которую мы передаём в качестве аргумента (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">michael</span>), никак не связано с именем параметра (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">bruce</span>). Не имеет значения, как называлось значение «дома» (в вызывающем коде); внутри <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print_twice</span> мы всех называем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">bruce</span>.


### **Функции, возвращающие значение, и функции без возвращаемого значения**

Некоторые функции, которые мы используем, например функции из модуля <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">math</span>, возвращают результат; за неимением более точного термина я называю их *функциями, возвращающими значение* (fruitful functions). Другие функции, такие как <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print_twice</span>, выполняют действие, но не возвращают значение. Их называют *функциями без возвращаемого значения* (void functions).

При вызове функции, возвращающей значение, почти всегда требуется использовать результат — например, присвоить его переменной или включить в состав выражения:



```python
x = math.cos(radians)
golden = (math.sqrt(5) + 1) / 2
```

При вызове функции в интерактивном режиме Python отображает результат:


```python
math.sqrt(5)
```

Однако в скрипте, если вы вызываете *функцию, возвращающую значение* (fruitful function) и не сохраняете её результат в переменной, возвращаемое значение просто исчезает без следа.


```python
result = print_twice('Bing')
print(result)
```

Значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span> — это не то же самое, что строка "None". Это специальное значение, имеющее собственный тип:


```python
print(type(None))
```

Чтобы вернуть результат из функции, используется инструкция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">return</span> внутри функции. Например, можно создать очень простую функцию с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">addtwo</span>, которая складывает два числа и возвращает результат.


```python
def addtwo(a, b):
    added = a + b
    return added

x = addtwo(3, 5)
print(x)
```

Когда этот скрипт выполняется, инструкция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span> выведет «8», поскольку функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">addtwo</span> была вызвана с аргументами 3 и 5. Внутри функции параметры <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">a</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">b</span> получили значения 3 и 5 соответственно.

Функция вычислила сумму этих двух чисел и сохранила её в локальной переменной функции с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">added</span>. Затем с помощью инструкции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">return</span> вычисленное значение было передано обратно в вызывающий код как результат функции. Этот результат был присвоен переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> и выведен на экран.


### **Зачем нужны функции?**

Не всегда очевидно, почему стоит разделять программу на функции. Однако для этого есть несколько причин:

- Создание новой функции даёт возможность дать имя группе инструкций, что делает программу более читаемой, понятной и удобной для отладки.

- Функции позволяют сократить объём программы за счёт устранения повторяющегося кода. Если позже потребуется внести изменения, достаточно сделать это в одном месте.

- Разделение большой программы на функции позволяет отлаживать её по частям, а затем объединять их в работающее целое.

- Хорошо спроектированные функции часто оказываются полезными в разных программах. После того как вы написали и отладили функцию, её можно повторно использовать.

В дальнейшем в книге определения функций часто будут использоваться для объяснения различных концепций. Важная часть навыка создания и использования функций заключается в том, чтобы функция корректно отражала некоторую идею, например «найти наименьшее значение в списке значений». Позже будет показан код, который находит минимальное значение в списке и представлен в виде функции с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">min</span>, принимающей список значений в качестве аргумента и возвращающей наименьшее значение из этого списка.


### **Отладка**

Если вы используете текстовый редактор для написания скриптов, могут возникнуть проблемы, связанные с пробелами и символами табуляции. Лучший способ избежать таких проблем — использовать только пробелы (без табуляции). Большинство редакторов, ориентированных на Python, по умолчанию используют пробелы, но не все.

Табуляции и пробелы обычно невидимы, из-за чего их сложно отлаживать, поэтому желательно использовать редактор, который автоматически управляет отступами.

Кроме того, не забывайте сохранять программу перед запуском. Некоторые среды разработки делают это автоматически, но не все. В противном случае программа, которую вы видите в редакторе, может отличаться от той, которая фактически выполняется.

Отладка может занять много времени, если вы снова и снова запускаете одну и ту же неверную версию программы.

Убедитесь, что код, который вы просматриваете, — это именно тот код, который выполняется. Если есть сомнения, добавьте в начало программы строку вроде <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print("hello")</span> и запустите её снова. Если вы не увидите <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">hello</span>, значит, выполняется не та программа.


### **Глоссарий**

**Алгоритм (algorithm)**

Общий способ решения определённого класса задач.

**Аргумент (argument)**

Значение, передаваемое функции при её вызове. Это значение присваивается соответствующующему параметру внутри функции.

**Тело функции (body)**

Последовательность инструкций внутри определения функции.

**Композиция (composition)**

Использование выражения как части более крупного выражения или инструкции как части более крупной инструкции.

**Детерминированный (deterministic)**

Свойство программы выполнять одни и те же действия при каждом запуске при одинаковых входных данных.

**Точечная нотация (dot notation)**

Синтаксис вызова функции из другого модуля путём указания имени модуля, затем точки и имени функции.

**Поток выполнения (flow of execution)**

Порядок, в котором инструкции выполняются во время работы программы.

**Функция, возвращающая значение (fruitful function)**

Функция, которая возвращает результат.

**Функция (function)**

Именованная последовательность инструкций, выполняющая некоторую полезную операцию. Функции могут принимать аргументы или не принимать их, а также могут возвращать результат или не возвращать его.

**Вызов функции (function call)**

Инструкция, которая выполняет функцию. Состоит из имени функции и списка аргументов.

**Определение функции (function definition)**

Инструкция, создающая новую функцию, задающая её имя, параметры и выполняемые инструкции.

**Объект функции (function object)**

Значение, создаваемое при определении функции. Имя функции — это переменная, которая ссылается на объект функции.

**Заголовок (header)**

Первая строка определения функции.

**Инструкция import (import statement)**

Инструкция, которая считывает файл модуля и создаёт объект модуля.

**Объект модуля (module object)**

Значение, создаваемое инструкцией import, предоставляющее доступ к данным и коду, определённым в модуле.

**Параметр (parameter)**

Имя, используемое внутри функции для обращения к значению, переданному в качестве аргумента.

**Псевдослучайный (pseudorandom)**

Относящийся к последовательности чисел, которые выглядят случайными, но генерируются детерминированной программой.

**Возвращаемое значение (return value)**

Результат работы функции. Если вызов функции используется как выражение, возвращаемое значение является значением этого выражения.

**Функция без возвращаемого значения (void function)**

Функция, которая не возвращает результат.

Перепишите вычисление заработной платы с учётом оплаты сверхурочных по ставке «полуторная оплата» и создайте функцию с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">computepay</span>, которая принимает два параметра (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">hours</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">rate</span>).


Enter Hours: 45

Enter Rate: 10

Pay: 475.0

```python
def computepay(hours, rate):
    if hours > 40:
        pay = 40 * rate + (hours - 40) * (rate * 1.5)
    else:
        pay = hours * rate
    return pay

try:
    hours = float(input("Enter Hours: "))
    rate = float(input("Enter Rate: "))
    print(f"Pay: {computepay(hours, rate)}")
except ValueError:
    print("Error, please enter numeric input.")
```

Запускайте программу многократно, чтобы проверить её работу для различных входных значений.

```python
def computegrade(score):
    if not (0.0 <= score <= 1.0):
        return "Score must be between 0.0 and 1.0!"
    if score >= 0.9:
        return 'A'
    elif score >= 0.8:
        return 'B'
    elif score >= 0.7:
        return 'C'
    elif score >= 0.6:
        return 'D'
    else:
        return 'F'

try:
    score = float(input("Enter a score: "))
    print(f"Grade: {computegrade(score)}")
except ValueError:
    print("Bad score")
```
