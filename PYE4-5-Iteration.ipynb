## **Итерации**

### **Обновление переменных**

Распространённый шаблон в операторах присваивания — это обновление переменной, при котором её новое значение зависит от предыдущего значения.


```python
x = x + 1
```

Это означает: "взять текущее значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span>, прибавить 1 и затем обновить <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span> новым значением".

Если попытаться обновить переменную, которая ещё не существует, возникнет ошибка, потому что Python сначала вычисляет правую часть выражения, а уже затем присваивает значение переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">x</span>:


```python
x = x + 1
```

Прежде чем обновлять переменную, её необходимо инициализировать, обычно с помощью простого присваивания:


```python
x = 0
x = x + 1
```

Обновление переменной путём прибавления 1 называется *инкрементом* (increment); вычитание 1 называется *декрементом* (decrement).


### **Оператор while**

Компьютеры часто используются для автоматизации повторяющихся задач. Повторение одинаковых или похожих действий без ошибок — это то, что компьютеры делают хорошо, а люди — плохо. Поскольку итерации встречаются очень часто, Python предоставляет несколько языковых конструкций, упрощающих их использование.

Одной из форм итерации в Python является оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span>. Ниже приведена простая программа, которая выполняет обратный отсчёт от пяти, а затем выводит "Blastoff!".


```python
n = 5
while n > 0:
    print(n)
    n = n - 1
print('Blastoff!')
```

Можно почти буквально читать оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span> как английское предложение. Это означает: "Пока <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">n</span> больше 0, выводи значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">n</span>, затем уменьшай его на 1. Когда значение станет равно 0, выйди из цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span> и выведи слово “Blastoff!”".

Более формально поток выполнения для оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span> выглядит так:

1. Вычисляется условие, результат — <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">True</span> или <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">False</span>.
2. Если условие ложно, выполнение цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span> прекращается, и программа переходит к следующей инструкции.
3. Если условие истинно, выполняется тело цикла, после чего управление возвращается к шагу 1.

Такой тип выполнения называется циклом, потому что на третьем шаге происходит возврат к началу. Каждый раз, когда выполняется тело цикла, называется *итерацией* (iteration). Для приведённого примера можно сказать: «цикл выполнился пять раз», то есть тело цикла было выполнено пять раз.

Тело цикла должно изменять значение одной или нескольких переменных так, чтобы в конечном итоге условие стало ложным и цикл завершился. Переменная, которая изменяется при каждой итерации и управляет завершением цикла, называется *переменной итерации* (iteration variable). Если такой переменной нет, цикл будет выполняться бесконечно, что приведёт к *бесконечному циклу* (infinite loop).


### **Бесконечные циклы**

Бесконечным источником шуток среди программистов является наблюдение, что инструкция на шампуне "Намылить, смыть, повторить" представляет собой бесконечный цикл, поскольку в ней нет переменной итерации, указывающей, сколько раз нужно выполнить повторение.

В случае с обратным отсчётом можно доказать, что цикл завершится: значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">n</span> конечно, и при каждой итерации оно уменьшается, поэтому в итоге обязательно станет равным 0. В других случаях цикл очевидно является бесконечным, поскольку в нём вообще отсутствует переменная итерации.

Иногда невозможно определить момент завершения цикла до тех пор, пока выполнение не дойдёт до середины его тела. В таком случае можно намеренно создать бесконечный цикл, а затем использовать оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">break</span>, чтобы выйти из него.

Этот цикл очевидно является бесконечным, потому что логическое выражение в операторе <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span> — это просто логическая константа <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">True</span>:


```python
n = 10
while True:
    print(n, end=' ')
    n = n - 1
print('Done!')
```

Если вы по ошибке запустите этот код, вы быстро научитесь останавливать "убежавший" процесс Python на своей системе или искать кнопку выключения компьютера. Эта программа будет выполняться бесконечно или до тех пор, пока не разрядится батарея, поскольку логическое выражение в начале цикла всегда истинно — оно равно константе <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">True</span>.

Хотя это пример некорректного бесконечного цикла, такой шаблон можно использовать для создания полезных циклов, если аккуратно добавить в тело цикла код, который явно завершает его с помощью оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">break</span> при достижении условия выхода.

Например, предположим, что вы хотите получать ввод от пользователя до тех пор, пока он не введёт <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">done</span>. Можно написать так:


```python
while True:
    line = input('> ')
    if line == 'done':
        break
    print(line)
print('Done!')
```

Условие цикла — <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">True</span>, то есть оно всегда истинно, поэтому цикл выполняется до тех пор, пока не встретит оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">break</span>.

На каждой итерации программа выводит приглашение в виде угловой скобки. Если пользователь вводит <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">done</span>, оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">break</span> завершает цикл. В противном случае программа выводит введённое пользователем значение и возвращается к началу цикла.

Вот пример выполнения:


```python
while True:
    line = input('> ')
    if line[0] == '#':
        continue
    if line == 'done':
        break
    print(line)
print('Done!')
```

Такой способ записи циклов <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span> является распространённым, поскольку позволяет проверять условие в любом месте цикла (а не только в начале) и формулировать условие завершения в утвердительной форме ("остановиться, когда это произойдёт"), а не в отрицательной ("продолжать, пока это не произойдёт").


### **Завершение итерации с помощью continue**

Иногда во время выполнения итерации цикла требуется завершить текущую итерацию и сразу перейти к следующей. В таком случае можно использовать оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">continue</span>, чтобы перейти к следующей итерации, не завершая выполнение оставшейся части тела цикла для текущей итерации.

Вот пример цикла, который повторяет ввод пользователя до тех пор, пока он не введёт "done", но игнорирует строки, начинающиеся с символа "#" (аналогично комментариям в Python):


```python
while True:
    line = input('> ')
    if line[0] == '#':
        continue
    if line == 'done':
        break
    print(line)
print('Done!')

# Code
```

Все строки выводятся, кроме той, которая начинается со знака "#", потому что при выполнении оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">continue</span> текущая итерация завершается, и управление возвращается к оператору <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span> для начала следующей итерации. Таким образом, инструкция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span> пропускается.


### **Определённые циклы с использованием for**

Иногда требуется выполнить цикл по набору элементов, например по списку слов, строкам файла или списку чисел. Когда у нас есть такой набор, можно использовать *определённый цикл* (definite loop) с помощью оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>.

Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span> называют *неопределённым циклом* (indefinite loop), потому что он выполняется до тех пор, пока некоторое условие не станет ложным. В отличие от него, цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> проходит по заранее известному набору элементов, поэтому количество итераций соответствует количеству элементов в этом наборе.

Синтаксис цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> похож на синтаксис цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span>: есть сам оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> и тело цикла:


```python
friends = ['Joseph', 'Glenn', 'Sally']
for friend in friends:
    print('Happy New Year:', friend)
print('Done!')
```

С точки зрения Python, переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">friends</span> — это список из трёх строк, и цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> проходит по этому списку, выполняя тело цикла один раз для каждой из трёх строк, что приводит к следующему выводу:


Happy New Year: Joseph

Happy New Year: Glenn

Happy New Year: Sally

Done!

Перевести этот цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> на "человеческий" язык не так просто, как цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span>, но если представить <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">friends</span> как набор, то смысл будет таким: "выполнять инструкции в теле цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> по одному разу для каждого друга из набора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">friends</span>".

Рассматривая цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>, слова <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">in</span> являются зарезервированными ключевыми словами Python, а <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">friend</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">friends</span> — это переменные.


```python
for friend in friends:
    print('Happy New Year:', friend)
```

В частности, переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">friend</span> является *переменной итерации* (iteration variable) для цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>. Значение переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">friend</span> изменяется на каждой итерации цикла и определяет момент его завершения.

Переменная итерации последовательно принимает значения трёх строк, хранящихся в переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">friends</span>.


### **Шаблоны циклов**

Часто мы используем цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> или <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span>, чтобы пройти по списку элементов или содержимому файла и найти, например, наибольшее или наименьшее значение среди данных.

Такие циклы обычно строятся следующим образом:

* Инициализация одной или нескольких переменных перед началом цикла
* Выполнение вычислений для каждого элемента в теле цикла с возможным изменением переменных
* Анализ полученных значений переменных после завершения цикла

Для демонстрации этих концепций и шаблонов будем использовать список чисел.

### **Циклы подсчёта и суммирования**

Например, чтобы посчитать количество элементов в списке, можно написать следующий цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>:


```python
count = 0
for itervar in [3, 41, 12, 9, 74, 15]:
    count = count + 1
print('Count: ', count)
```

Перед началом цикла мы устанавливаем переменную <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">count</span> в ноль, затем пишем цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>, который проходит по списку чисел. Переменная итерации называется <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">itervar</span>, и хотя мы не используем её напрямую в теле цикла, именно она управляет выполнением цикла, заставляя тело выполняться один раз для каждого значения в списке.

В теле цикла мы прибавляем 1 к текущему значению <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">count</span> для каждого элемента списка. Пока цикл выполняется, значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">count</span> показывает, сколько элементов мы уже "обработали на данный момент".

После завершения цикла значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">count</span> становится общим количеством элементов. Итоговое значение как бы «само получается» к концу выполнения цикла. Мы строим цикл так, чтобы по его завершении получить нужный результат.

Другой похожий цикл, который вычисляет сумму набора чисел, выглядит следующим образом:


```python
total = 0
for itervar in [3, 41, 12, 9, 74, 15]:
    total = total + itervar
print('Total: ', total)
```

В этом цикле мы уже используем переменную итерации. Вместо того чтобы просто прибавлять 1 к <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">count</span>, как в предыдущем примере, мы добавляем само значение (3, 41, 12 и т. д.) к накопленной сумме на каждой итерации цикла.

Если рассмотреть переменную <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">total</span>, то она содержит «текущую накопленную сумму значений на данный момент». До начала цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">total</span> равна нулю, поскольку ни одно значение ещё не обработано; во время выполнения цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">total</span> хранит промежуточную сумму; а после завершения цикла — итоговую сумму всех элементов списка.

По мере выполнения цикла переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">total</span> накапливает сумму элементов; переменная, используемая таким образом, иногда называется *накопителем* (accumulator).

Ни цикл подсчёта, ни цикл суммирования на практике не так полезны, поскольку существуют встроенные функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len()</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sum()</span>, которые соответственно вычисляют количество элементов в списке и их сумму.


### **Циклы поиска максимума и минимума**

Чтобы найти наибольшее значение в списке или последовательности, можно построить следующий цикл:


```python
largest = None
print('Before:', largest)
for itervar in [3, 41, 12, 9, 74, 15]:
    if largest is None or itervar > largest :
        largest = itervar
    print('Loop:', itervar, largest)
print('Largest:', largest)
```

Переменную <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">largest</span> удобно рассматривать как «наибольшее значение, которое мы встретили на данный момент». Перед началом цикла мы присваиваем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">largest</span> значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span>. Это специальная константа, которую можно использовать для обозначения «пустого» значения переменной.

До начала цикла наибольшее значение равно <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span>, поскольку мы ещё не обработали ни одного элемента. Во время выполнения цикла, если <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">largest</span> равно <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span>, мы принимаем первое встреченное значение как текущее максимальное. Например, на первой итерации, когда значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">itervar</span> равно 3, переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">largest</span> сразу становится равной 3.

После первой итерации <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">largest</span> уже не равна <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span>, поэтому вторая часть составного условия (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">itervar > largest</span>) срабатывает только тогда, когда встречается значение больше текущего максимума. Когда появляется новое «ещё большее» значение, мы обновляем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">largest</span>. В результате выполнения программы видно, как значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">largest</span> изменяется с 3 до 41, затем до 74.

В конце цикла мы просматриваем все элементы, и переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">largest</span> содержит наибольшее значение в списке.

Чтобы найти наименьшее значение, код будет почти таким же, с небольшим изменением:


```python
smallest = None
print('Before:', smallest)
for itervar in [3, 41, 12, 9, 74, 15]:
    if smallest is None or itervar < smallest:
        smallest = itervar
    print('Loop:', itervar, smallest)
print('Smallest:', smallest)
```

Снова переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">smallest</span> представляет собой "наименьшее значение на данный момент" — до, во время и после выполнения цикла. Когда цикл завершён, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">smallest</span> содержит минимальное значение в списке.

Как и в случае с подсчётом и суммированием, встроенные функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">max()</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">min()</span> делают написание подобных циклов необязательным.

Ниже приведена простая версия встроенной функции Python <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">min()</span>:


```python
def min(values):
    smallest = None
    for value in values:
        if smallest is None or value < smallest:
            smallest = value
    return smallest
```

В версии функции для поиска наименьшего значения мы удалили все инструкции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span>, чтобы сделать её эквивалентной встроенной функции Python <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">min()</span>.


### **Отладка**

По мере написания более крупных программ вы можете заметить, что начинаете тратить больше времени на отладку. Чем больше кода, тем выше вероятность ошибок и тем больше мест, где могут скрываться баги.

Один из способов сократить время отладки — это *отладка методом деления пополам* (debugging by bisection). Например, если в программе 100 строк и проверять их по одной, потребуется 100 шагов.

Вместо этого попробуйте разделить проблему пополам. Посмотрите на середину программы или близкую к ней часть и найдите промежуточное значение, которое можно проверить. Добавьте инструкцию <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span> (или что-то другое с проверяемым эффектом) и запустите программу.

Если проверка в середине даёт неверный результат, значит ошибка находится в первой половине программы. Если результат верный — ошибка во второй половине.

Каждый раз, выполняя такую проверку, вы сокращаете область поиска вдвое. Теоретически, уже через шесть шагов (что значительно меньше 100) можно сузить проблему до одной-двух строк кода.

На практике не всегда очевидно, где находится «середина программы», и не всегда возможно её проверить. Нет смысла буквально считать строки и искать точный центр. Вместо этого стоит определить участки программы, где вероятнее всего могут быть ошибки, и места, где удобно выполнить проверку. Затем выбрать точку, в которой вероятность того, что ошибка находится до неё или после неё, примерно одинакова.


### **Глоссарий**

**Накопитель (accumulator)**

Переменная, используемая в цикле для накопления результата.

**Счётчик (counter)**

Переменная, используемая в цикле для подсчёта количества повторений некоторого события. Обычно счётчик инициализируется нулём, а затем увеличивается при каждом нужном событии.

**Декремент (decrement)**

Обновление, уменьшающее значение переменной.

**Инициализация (initialize)**

Присваивание начального значения переменной, которое затем будет изменяться.

**Инкремент (increment)**

Обновление, увеличивающее значение переменной (часто на единицу).

**Бесконечный цикл (infinite loop)**

Цикл, в котором условие завершения никогда не выполняется или вовсе отсутствует.

**Итерация (iteration)**

Повторное выполнение набора инструкций с использованием либо функции, вызывающей саму себя, либо цикла.

**Упражнение 1:**
Напишите программу, которая многократно считывает целые числа до тех пор, пока пользователь не введёт <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">done</span>. После ввода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">done</span> выведите сумму, количество и среднее значение введённых чисел.

Если пользователь вводит что-то отличное от целого числа, необходимо обработать ошибку с помощью <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">try</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">except</span>, вывести сообщение об ошибке и перейти к следующему вводу.


Enter a number: 4

Enter a number: 5

Enter a number: bad data

Invalid input

Enter a number: 7

Enter a number: done

16 3 5.333333333333333

```python
total = 0
count = 0
while True:
    user_input = input("Enter a number: ")
    if user_input == "done":
        break
    try:
        number = int(user_input)
    except ValueError:
        print("Invalid input")
        continue
    total += number
    count += 1

if count > 0:
    average = total / count
    print(f"{total} {count} {average}")
else:
    print("No numbers were entered.")
```

**Упражнение 2:**
Напишите ещё одну программу, которая, как и в предыдущем задании, запрашивает у пользователя список чисел, а в конце выводит максимальное и минимальное значения вместо среднего.


```python
largest = None
smallest = None

while True:
    user_input = input("Enter a number: ")

    if user_input == "done":
        return false

    try:
        number = int(user_input)
    except ValueError:
        print("Invalid input")
        continue

    if largest is None or number > largest:
        largest = number

    if smallest is None or number < smallest:
        smallest = number

print(f"Maximum is {largest}")
print(f"Minimum is {smallest}")
```
