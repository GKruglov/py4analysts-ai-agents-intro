## **Строки**
### **Строка – это последовательность**
Строка — это _последовательность_ символов. Вы можете получить доступ к символам по одному с помощью оператора скобок:


```python
fruit = 'banana'
letter = fruit[1]
```

Второе выражение извлекает символ с индексом 1 из переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">fruit</span> и присваивает его переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">letter</span>.

Выражение в скобках называется индексом. Индекс указывает, какой символ в последовательности вам нужен (отсюда и название).

Но вы можете получить не то, что ожидаете:


```python
print(letter)
```

Для большинства людей первая буква слова banana — это «b», а не «a». Но в Python индекс представляет собой смещение от начала строки, и смещение первой буквы равно нулю.

```python
letter = fruit[0]
print(letter)
```

Таким образом, "b" — это 0-я буква (нолевая) слова banana, "a" — 1-я буква (первая), а "n" — 2-я (вторая) буква.

### **Получение длины строки с помощью <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len</span>**
<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len</span> — встроенная функция, которая возвращает количество символов в строке:


```python
fruit = 'banana'
len(fruit)
```

Чтобы получить последнюю букву строки, можно попробовать сделать что-то вроде этого:

```python
length = len(fruit)
#last = fruit[length]
#IndexError: string index out of range
```

Причина <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">IndexError</span> заключается в том, что в слове banana нет буквы с индексом 6. Поскольку мы начали отсчет с нуля, шесть букв пронумерованы от 0 до 5. Чтобы получить последний символ, нужно вычесть 1 из длины (<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">length</span>):

```python
last = fruit[length - 1]
print(last)
```

В качестве альтернативы можно использовать отрицательные индексы, которые считают в обратном порядке от конца строки. Выражение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">fruit[-1]</span> возвращает последнюю букву, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;"> fruit[-2]</span> — предпоследнюю и т. д.

### **Обход строки с помощью цикла**

Многие вычисления включают в себя обработку строки по одному символу за раз. Часто они начинаются с начала, выбирают каждый символ по очереди, выполняют над ним какое-либо действие и продолжаются до конца. Такой паттерн обработки называется обходом. Один из способов написания обхода — с помощью цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span>:

```python
index = 0
while index < len(fruit):
    letter = fruit[index]
    print(letter)
    index = index + 1
```

Этот цикл проходит по строке и выводит каждую букву в отдельной строке. Условие цикла — <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">index < len(fruit)</span>, поэтому когда <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">index</span> равен длине строки, условие становится ложным, и тело цикла не выполняется. Последний доступный символ — это символ с индексом <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len(fruit)-1</span>, который является последним символом в строке.

**Упражнение 1:** Напишите цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span>, который начинается с последнего символа в строке и проходит в обратном направлении до первого символа в строке, выводя каждую букву в отдельной строке, но в обратном порядке.

```python
index = len(fruit)
while index > 0:
    letter = fruit[index - 1]
    print(letter)
    index -= 1
```

Другой способ написать обход — с помощью цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>:

```python
for char in fruit:
    print(char)
```

Каждый раз при прохождении цикла следующий символ в строке присваивается переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">char</span>. Цикл продолжается до тех пор, пока не останется символов.

### **Фрагменты строк (String slices)**
Фрагмент строки называется _срезом_. Выбор среза аналогичен выбору символа:


```python
s = 'Monty Python'
print(s[0:5])
print(s[6:12])
```

Оператор [n:m] возвращает часть строки от "n-го" символа до "m-го" символа, включая первый, но исключая последний.

Если вы опустите первый индекс (перед двоеточием), фрагмент начнется с начала строки. Если вы опустите второй индекс, фрагмент будет до конца строки:


```python
fruit = 'banana'
print(fruit[:3])
print(fruit[3:])
```

Если первый индекс больше или равен второму, результатом будет _пустая строка_, представленная двумя кавычками:

```python
fruit = 'banana'
fruit[3:3]
```

Пустая строка не содержит символов и имеет длину 0, но в остальном она такая же, как и любая другая строка.

### **Строки являются неизменяемыми**
Заманчиво использовать оператор в левой части присваивания с целью изменения символа в строке. Например:


```python
greeting = 'Hello, world!'
greeting[0] = 'J'
TypeError: 'str' object does not support item assignment
```

В данном случае "объект" — это строка, а "элемент" — символ, который вы пытались присвоить. На данный момент _объект_ — это то же самое, что и значение, но позже мы уточним это определение. _Элемент_ — это одно из значений в последовательности.


Причина ошибки заключается в том, что строки являются _неизменяемыми_, то есть вы не можете изменить существующую строку. Лучшее, что вы можете сделать, — это создать новую строку, которая будет вариацией исходной:


```python
greeting = 'Hello, world!'
new_greeting = 'J' + greeting[1:]
print(new_greeting)
```

В этом примере новая первая буква добавляется к срезу строки <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">greeting</span>. При этом исходная строка не изменяется.

### **Циклическое повторение и подсчет (Looping and counting)**
Следующая программа подсчитывает количество вхождений буквы "a" в строке:


```python
word = 'banana'
count = 0
for letter in word:
    if letter == 'a':
        count = count + 1
print(count)
```

Эта программа демонстрирует другой паттерн вычислений, называемый _счетчиком_. Переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">count</span> инициализируется значением 0, а затем инкрементируется каждый раз, когда встречается буква "a". При выходе из цикла count содержит результат: общее количество букв "a".

**Упражнение 3:** Закапсулируйте этот код в функцию с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">count</span> и обобщите ее так, чтобы она принимала строку и букву в качестве аргументов.

```python
def count(word, letter):
    count = 0
    for char in word:
        if char == letter:
            count += 1
    return count
  
result = count('banana', 'a')
print(result)
```

### **Сравнение строк**
Операторы сравнения работают со строками. Чтобы проверить, равны ли две строки:


```python
if word == 'banana':
    print('All right, bananas.')
```

Другие операции сравнения полезны для упорядочивания слов в алфавитном порядке:

```python
if word < 'banana':
    print('Your word,' + word + ', comes before banana.')
elif word > 'banana':
    print('Your word,' + word + ', comes after banana.')
else:
    print('All right, bananas.')
```

Python обрабатывает заглавные и строчные буквы не так, как люди. Все заглавные буквы стоят перед всеми строчными буквами, поэтому:



```python
#Your word, Pineapple, comes before banana.
```

Обычным способом решения этой проблемы является преобразование строк в стандартный формат, например, в нижний регистр, перед выполнением сравнения. Имейте это в виду на случай, если вам придется защищаться от человека, вооруженного ананасом.

### **Методы строк**
Строки являются примером _объектов_ Python. Объект содержит как данные (саму строку), так и _методы_, которые фактически являются функциями, встроенными в объект и доступными для любого _экземпляра_ объекта.

В Python есть функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dir</span>, которая выводит список методов, доступных для объекта. Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">type</span> показывает тип объекта, а функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dir</span> — доступные методы.


```python
stuff = 'Hello world'
type(stuff)
```

```python
dir(stuff)
```

```python
help(str.capitalize)
```

Хотя функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dir</span> выводит список методов, а с помощью команды <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">help</span> можно получить простую документацию по методу, лучшим источником документации по методам работы со строками является
https://docs.python.org/library/stdtypes.html#string-methods.

Вызов _метода_ похож на вызов функции (он принимает аргументы и возвращает значение), но синтаксис отличается. Мы вызываем метод, добавляя имя метода к имени переменной с помощью точки в качестве разделителя.

Например, метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">upper</span> принимает строку и возвращает новую строку, в которой все буквы заглавные:
Вместо синтаксиса функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">upper(word)</span> используется синтаксис метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">word.upper()</span>.


```python
word = 'bAnAnA'
new_word = word.upper()
print(new_word)
```

Эта форма точечной нотации указывает имя метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">upper</span> и имя строки, к которой будет применен метод, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">word</span>. Пустые скобки означают, что этот метод не принимает аргументов.

Вызов метода называется _invocation (инвокация)_; в данном случае мы говорим, что инвоцируем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">upper</span> на <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">word</span>.

Например, существует метод строки <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span>, который ищет положение одной строки внутри другой:


```python
word = 'banana'
index = word.find('a')
print(index)
```

В этом примере мы вызываем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span> на <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">word</span> и передаем искомую букву в качестве параметра.

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span> может находить как подстроки, так и символы:


```python
word.find('na')
```

### **Разбор (parsing) строк**
Часто нам нужно проанализировать строку и найти в ней подстроку. Например, если нам представлена серия строк, отформатированных следующим образом:

<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">From stephen.marquard@ uct.ac.za Sat Jan  5 09:14:16 2008</span>

и мы хотим извлечь только вторую половину адреса (т. е. <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">uct.ac.za</span>) из каждой строки, мы можем сделать это с помощью метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span> и разбиения строки.
Сначала мы найдем положение знака @ в строке. _Затем_ мы найдем положение первого пробела после знака @. И затем мы будем использовать разбиение строки, чтобы извлечь часть строки, которую мы ищем.


```python
data = 'From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008'
atpos = data.find('@')
print(atpos)

sppos = data.find(' ',atpos)
print(sppos)

host = data[atpos+1:sppos]
print(host)
```

Мы используем версию метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span>, которая позволяет нам указать позицию в строке, с которой мы хотим начать поиск. При выделении мы извлекаем символы от "одного за знаком @ до символа пробела, но _не включая его_".

Документация по методу <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span> доступна по адресу

https://docs.python.org/library/stdtypes.html#string-methods.


### **Форматированные строковые литералы**
Форматированный строковый литерал (часто называемый просто f-строкой) позволяет использовать выражения Python в строковых литералах. Это достигается путем добавления буквы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">f</span> перед строковым литералом и заключения выражений в фигурные скобки <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">{}</span>.
Например, заключение имени переменной в фигурные скобки внутри f-строки приведет к его замене на его значение:


```python
camels = 42
f'{camels}'
```

Результатом будет строка "42", которую не следует путать с целым числом 42.
Выражение может появляться в любом месте строки, поэтому вы можете встроить значение в предложение:


```python
camels = 42
f'I have spotted {camels} camels.'
```

В одну строковую литерал можно включить несколько выражений, чтобы создать более сложные строки.

```python
years = 3
count = .1
species = 'camels'
f'In {years} years I have spotted {count} {species}.'
```

Форматированные строковые литералы обладают широкими возможностями и могут делать гораздо больше, чем описано здесь. Подробнее о них можно прочитать по адресу

https://docs.python.org/3/tutorial/inputoutput.html#formatted-string-literals.

Например, посмотрите на программу, которую мы использовали для демонстрации цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span> в главе, посвященной итерации:


```python
while True:
    line = input('> ')
    if line[0] == '#':
        continue
    if line == 'done':
        break
    print(line)
print('Done!')
```

Посмотрите, что происходит, когда пользователь вводит пустую строку:

```python
#> hello there
#hello there
#> # don't print this
#> print this!
#print this!
#>
#Traceback (most recent call last):
#  File "copytildone.py", line 3, in <module>
#    if line[0] == '#':
#IndexError: string index out of range
```

Код работает нормально, пока не встречается пустая строка. В этом случае нулевой символ отсутствует, поэтому мы получаем трассировку. Есть два решения, которые позволяют сделать третью строку "безопасной", даже если она пустая.

Один из вариантов — просто использовать метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">startswith</span>, который возвращает <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">False</span>, если строка пустая.

```python
if len(line) > 0 and line[0] == '#':
```

Другой способ — безопасно написать оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span>, используя паттерн _guardian_, и убедиться, что второе логическое выражение вычисляется только в том случае, если в строке есть хотя бы один символ:

```python
if len(line) > 0 and line[0] == '#':
```

### **Глоссарий**

**counter (счётчик)**

Переменная, используемая для подсчёта чего-либо. Обычно инициализируется значением 0, после чего её значение увеличивается.

**empty string (пустая строка)**

Строка без символов и длиной 0, представленная двумя кавычками.

**flag (флаг)**

Булева переменная, используемая для указания, является ли условие истинным или ложным.

**invocation (вызов)**

Инструкция (оператор), которая вызывает метод.

**immutable (неизменяемый)**

Свойство последовательности, элементы которой нельзя изменять или переназначать.

**index (индекс)**

Целое число, используемое для выбора элемента в последовательности, например символа в строке.

**item (элемент)**

Одно из значений, входящих в последовательность.

**method (метод)**

Функция, связанная с объектом и вызываемая с использованием точечной нотации.

**object (объект)**

То, на что может ссылаться переменная. На данный момент термины «объект» и «значение» можно использовать как взаимозаменяемые.

**search (поиск)**

Шаблон обхода, который прекращается, когда найдено искомое значение.

**sequence (последовательность)**

Упорядоченное множество, то есть множество значений, в котором каждое значение идентифицируется целочисленным индексом.

**slice (срез)**

Часть строки, определяемая диапазоном индексов.

**traverse (обход)**

Итерация по элементам последовательности с выполнением одинаковой операции для каждого элемента.

**Упражнение 5:** Разделение строк
Возьмите следующий код Python, который хранит строку:

<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">str = „X-DSPAM-Confidence: 0.8475“</span>

Используйте <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span> и разделение строк, чтобы извлечь часть строки после символа двоеточия, а затем используйте функцию <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">float</span>, чтобы преобразовать извлеченную строку в число с плавающей запятой.


```python
string = 'X-DSPAM-Confidence: 0.8475'

position = string.find(':')

numeric_part = string[position + 1:]

float_value = float(numeric_part.strip())

print(float_value)
```

**Упражнение 6:** Методы строк

Прочитайте документацию по методам строк по адресу

https://docs.python.org/library/stdtypes.html#string-methods.

Вы можете поэкспериментировать с некоторыми из них, чтобы убедиться, что понимаете, как они работают. Особенно полезны методы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">strip</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">replace</span>.

В документации используется синтаксис, который может вызвать путаницу. Например, в <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find(sub[, start[, end]])</span> скобки обозначают необязательные аргументы. Таким образом, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sub</span> является обязательным, а <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">start</span> — необязательным, и если вы включаете <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">start</span>, то end становится необязательным.


```python
s = "   Hello World!   "
stripped = s.strip()
print(stripped)  

custom_strip = "!!HellO!!".strip("!")
print(custom_strip)  
left_trimmed = "---example---".lstrip("-")
right_trimmed = "---example---".rstrip("-")
print(left_trimmed)
print(right_trimmed)
sentence = "hello hello hello"
replaced_once = sentence.replace("hello", "hi", 1)
all_replaced = sentence.replace("hello", "hi")

print(replaced_once)  
print(all_replaced)   
words = "apple banana cherry".split()
fruits = "apple-banana-cherry".split('-')
limited_split = "one-two-three-four-five".split('-', 2)

print(words)          
print(fruits)        
print(limited_split) 
lowercase = "HELLO WORLD!".lower()
uppercase = "hello world!".upper()
title_case = "this is a test".title()

print(lowercase)     
print(uppercase)     
print(title_case)
starts_with_h = "hello".startswith("h")      
ends_with_exclamation = "Wow!".endswith("!") 
substring_check = "abcdefghi".startswith("cd", 2)
print(starts_with_h)
print(ends_with_exclamation)
print(substring_check)
items = ["apples", "oranges", "pears"]
joined_string = ", ".join(items)
print(joined_string)
```
