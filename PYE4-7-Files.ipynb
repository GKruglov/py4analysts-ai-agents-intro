### **Чтение файлов**

Хотя файловый дескриптор (*file handle*) не содержит сами данные файла, довольно просто написать цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>, чтобы пройти по всем строкам файла и посчитать их количество:

````python
fhand = open('mbox-short.txt')
count = 0
for line in fhand:
    count = count + 1
print('Line Count:', count)
````

Мы можем использовать файловый дескриптор как последовательность в цикле <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>. Этот цикл просто считает количество строк в файле и выводит результат. Упрощённо это можно перевести так: «для каждой строки в файле, представленном файловым дескриптором, увеличь переменную счётчика <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">count</span>на 1».

Причина, по которой функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">open</span> не читает весь файл сразу, заключается в том, что файл может быть очень большим — с данными в несколько гигабайт. Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">open</span> выполняется за одинаковое время независимо от размера файла. Само чтение данных происходит уже при выполнении цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>.

Когда файл читается таким образом, Python сам разбивает данные на строки с помощью символа новой строки. Python читает каждую строку до символа новой строки и включает этот символ в конец переменной строки на каждой итерации цикла.

Поскольку цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> читает данные по одной строке за раз, он может эффективно обрабатывать и подсчитывать строки даже в очень больших файлах, не занимая много оперативной памяти. Программа выше может подсчитать строки в файле любого размера, используя очень мало памяти, потому что каждая строка читается, обрабатывается и затем удаляется из памяти.

Если вы знаете, что файл относительно небольшой по сравнению с объёмом оперативной памяти, можно прочитать весь файл сразу в одну строку, используя метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">read</span>:

````python
fhand = open('mbox-short.txt')
inp = fhand.read()
print(len(inp))
print(inp[:20])
````

В этом примере всё содержимое (все 94 626 символов) файла *mbox-short.txt* читается напрямую в переменную <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">inp</span>. Мы используем срез строки, чтобы вывести первые 20 символов данных, хранящихся в <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">inp</span>.

Когда файл читается таким способом, все символы, включая все строки и символы новой строки, становятся одной большой строкой в переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">inp</span>. Хорошей практикой является сохранять результат <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">read</span> в переменную, потому что каждый вызов <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">read</span> «исчерпывает» данные:

````python
fhand = open('mbox-short.txt')
print(len(fhand.read()))
print(len(fhand.read()))
````

Помните, что такой способ использования функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">open</span> стоит применять только тогда, когда объём данных файла комфортно помещается в оперативную память компьютера. Если файл слишком большой, следует писать программу так, чтобы читать его по частям — с помощью циклов <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> или <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">while</span>.

### **Поиск в файле**

Когда вы ищете данные в файле, очень распространённый подход — читать файл, игнорируя большую часть строк и обрабатывая только те строки, которые соответствуют определённому условию. Мы можем объединить чтение файла с методами строк, чтобы создать простые механизмы поиска.

Например, если мы хотим прочитать файл и выводить только те строки, которые начинаются с префикса «From:», можно использовать строковый метод *startswith*, чтобы отбирать только нужные строки. Когда программа выполняется, мы получаем следующий вывод:

````python
fhand = open('mbox-short.txt')
for line in fhand:
    if line.startswith('From:'):
        print(line)
````

Вывод выглядит хорошо, потому что мы видим только строки, начинающиеся с «From:». Но почему появляются лишние пустые строки? Это связано с невидимым символом новой строки. Каждая строка уже заканчивается *символом новой строки*, поэтому <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span> выводит *строку*, содержащую этот символ, а затем сам <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span> добавляет *ещё один* перевод строки — в итоге получается эффект двойного интервала.

Мы могли бы использовать срез строки, чтобы не выводить последний символ, но проще использовать метод rstrip, который удаляет пробельные символы справа. Результат работы программы:

````python
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.rstrip()
    if line.startswith('From:'):
        print(line)
````

По мере того как ваши программы обработки файлов становятся сложнее, вы можете структурировать циклы поиска с использованием <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">continue</span>. Основная идея цикла поиска — находить «интересные» строки и фактически пропускать «неинтересные». А когда находится интересная строка — выполнять с ней какие-то действия.

Можно структурировать цикл пропуском неинтересных строк по следующему шаблону:

````python
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.rstrip()
    # Skip 'uninteresting lines'
    if not line.startswith('From:'):
        continue
    # Process our 'interesting' line
    print(line)
````

Результат программы будет таким же. Проще говоря, неинтересные строки — это те, которые не начинаются с «From:», и мы пропускаем их с помощью <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">continue</span>. Для «интересных» строк (то есть начинающихся с «From:») мы выполняем обработку.

Мы также можем использовать метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span>, чтобы имитировать поиск в текстовом редакторе, находящий строки, где искомая подстрока встречается в любом месте строки. Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">find</span> ищет вхождение подстроки в строке и возвращает её позицию или -1, если подстрока не найдена. С его помощью можно написать такой цикл, чтобы находить строки, содержащие «@uct.ac.za» (то есть письма из Университета Кейптауна в Южной Африке). Это даёт следующий вывод:

````python
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.rstrip()
    if line.find('@uct.ac.za') == -1: continue
    print(line)
````

Здесь также используется сокращённая форма оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span>, где <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">continue</span> записан в одной строке с условием. Эта запись работает так же, как если бы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">continue</span> находился на следующей строке с отступом.

### **Позволим пользователю выбирать имя файла**

На самом деле нам не хочется каждый раз редактировать наш код на Python, когда нужно обработать другой файл. Гораздо удобнее попросить пользователя вводить имя файла при каждом запуске программы, чтобы он мог использовать её с разными файлами без изменения кода.

Сделать это довольно просто — можно считать имя файла от пользователя с помощью <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">input</span>:

````python
# search6.py
fname = input('Enter the file name: ')
fhand = open(fname)
count = 0
for line in fhand:
    if line.startswith('Subject:'):
        count = count + 1
print('There were', count, 'subject lines in', fname)
````

Мы считываем имя файла от пользователя и сохраняем его в переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">fname</span>, после чего открываем этот файл. Теперь программу можно запускать много раз с разными файлами.

Пример запуска:

````python
fname = input('Enter the file name: ')
try:
    fhand = open(fname)
except:
    print('File cannot be opened:', fname)
    exit()
count = 0
for line in fhand:
    if line.startswith('Subject:'):
        count = count + 1
print('There were', count, 'subject lines in', fname)
````

Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">exit</span> завершает программу. Это функция, которую мы вызываем, и она не возвращает управление. Теперь, когда пользователь (или команда QA) вводит что-то некорректное или неправильное имя файла, мы «перехватываем» ошибку и корректно обрабатываем её:

```python
python search7.py

```

Защита вызова <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">open</span> — хороший пример правильного использования <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">try</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">except</span> в Python. Когда мы делаем что-то «по-питоновски», мы говорим, что это «Pythonic» подход. Можно сказать, что приведённый выше пример — это именно такой «питоновский» способ открытия файла.

Когда ты станешь более опытным в Python, ты сможешь обсуждать с другими разработчиками, какое из двух одинаковых решений является «более Pythonic». Стремление быть «более Pythonic» отражает идею о том, что программирование — это не только инженерия, но и искусство. Мы хотим не просто чтобы код работал, но чтобы он был элегантным и воспринимался как элегантный другими разработчиками.

### **Запись в файлы**

Чтобы записывать в файл, его нужно открыть в режиме "w" (write) — второй параметр функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">open</span>:

````python
fout = open('output.txt', 'w')
print(fout)
````

Если файл уже существует, открытие в режиме записи очистит все старые данные и начнёт запись с нуля, так что будь осторожен. Если файла не существует, будет создан новый.

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">write</span> объекта файлового дескриптора записывает данные в файл и возвращает количество записанных символов. По умолчанию используется текстовый режим записи строк.

````python
line1 = "This here's the wattle,\n"
fout.write(line1)
````

Файловый объект сам отслеживает текущее положение в файле, поэтому при следующем вызове <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">write</span> новые данные будут добавлены в конец.

При записи в файл нужно самостоятельно управлять концами строк — явно добавлять символ новой строки, когда нужно закончить строку. Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span> автоматически добавляет перевод строки, а <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">write</span> — нет.

````python
line2 = 'the emblem of our land.\n'
fout.write(line2)
````

Когда вы заканчиваете запись, необходимо закрыть файл, чтобы гарантировать, что последние данные будут физически записаны на диск и не потеряются при отключении питания.

````python
fout.close()
````

Мы могли бы закрывать и файлы, которые открываем только для чтения, но можно немного расслабиться, если открываем немного файлов — Python сам закроет все открытые файлы при завершении программы. Однако при записи файлов лучше всегда явно закрывать их, чтобы ничего не оставить на волю случая.

### **Отладка**

Когда вы читаете и записываете файлы, можно столкнуться с проблемами, связанными с пробельными символами. Такие ошибки сложно отлаживать, потому что пробелы, табуляции и символы новой строки обычно невидимы:

````python
s = '1 2\t 3\n 4'
print(s)
````

Встроенная функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">repr</span> может помочь. Она принимает любой объект в качестве аргумента и возвращает его строковое представление. Для строк она отображает пробельные символы с помощью escape-последовательностей:

````python
print(repr(s))
````

Это может быть полезно при отладке.

Ещё одна проблема, с которой можно столкнуться, — это то, что разные системы используют разные символы для обозначения конца строки. Некоторые системы используют символ новой строки (
<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">\n</span>), другие — символ возврата каретки (
<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">\r</span>), а некоторые — оба сразу. Если вы переносите файлы между разными системами, такие различия могут вызывать проблемы.

Для большинства систем существуют приложения, которые позволяют конвертировать один формат в другой. Подробнее об этом можно узнать по ссылке: https://www.wikipedia.org/wiki/Newline
. Или, конечно, можно написать такое приложение самостоятельно.

**Упражнение 2**:
Напишите программу, которая запрашивает имя файла, затем читает файл и ищет строки вида:
```python
X-DSPAM-Confidence: 0.8475
```

Когда встречается строка, начинающаяся с X-DSPAM-Confidence:, нужно извлечь из неё число с плавающей точкой. Подсчитайте такие строки и вычислите сумму значений spam-confidence. В конце файла выведите среднее значение spam-confidence.
Пример работы:

```python
Enter the file name: mbox.txt
Average spam confidence: 0.894128046745

Enter the file name: mbox-short.txt
Average spam confidence: 0.750718518519
```
Проверьте программу на файлах mbox.txt и mbox-short.txt.

````python
def Avg_confidence(file_name=None):
    if file_name is None:
        file_name = input("Enter the file name: ")

    print(f"Enter the file name: {file_name}")
    count = 0
    total = 0
    try:
        with open(file_name, "r") as file:
            for line in file:
                if line.startswith("X-DSPAM-Confidence:"):
                    colon_pos = line.find(":")

                    number_str = line[colon_pos + 1 :].strip()
                    confidence = float(number_str)

                    total += confidence
                    count += 1

    except FileNotFoundError:
        print(f"File not found: {file_name}")

    if count > 0:
        average = total / count
        print(f"Average spam confidence: {average}")
    else:
        print("No spam confidence lines found in the file.")


Avg_confidence("mbox.txt")
Avg_confidence("mbox-short.txt")
````

**Упражнение 3**: Иногда программисты, когда им скучно или просто ради шутки, добавляют в программы безобидные «пасхалки» (Easter Egg). Измените программу, которая запрашивает имя файла, так, чтобы она выводила смешное сообщение, если пользователь введёт точное имя файла "na na boo boo". Во всех остальных случаях (если файл существует или не существует) программа должна работать как обычно.

Пример работы программы:

```python
python egg.py
Enter a file name: mbox.txt
There were 1797 subject lines in mbox.txt

python egg.py
Enter a file name: missing.tyxt

python egg.py
Enter a file name: na na boo boo
NA NA BOO BOO TO YOU - You have been punk'd!
```
Мы не призываем добавлять пасхалки в свои программы — это просто упражнение.

````python
def Prunk(file_name=None):
    if file_name is None:
        file_name = input("Enter the file name: ")
    elif file_name == "na na boo boo":
        print("NA NA BOO BOO TO YOU - You have been punk'd!")
    else:
        count = 0
        try:
            with open(file_name, "r") as file:
                for line in file:
                    if line.startswith("Subject:"):
                        count += 1

            print(f"There were {count} subject lines in {file_name}")

        except FileNotFoundError:
            print(f"File cannot be opened: {file_name}")


Prunk("mbox.txt")
Prunk("missing.tyxt")
Prunk("na na boo boo")
````
