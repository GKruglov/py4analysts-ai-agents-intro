## **Списки**

### **Список — это последовательность**

Подобно строке, список является *последовательностью* (sequence) значений. В строке значениями являются символы; в списке они могут быть любого типа. Значения в списках называются *элементами* (elements) или иногда *элементами списка* (items).

Существует несколько способов создать новый список; самый простой — заключить элементы в квадратные скобки («[» и «]»):

```python
[10, 20, 30, 40]
['crunchy frog', 'ram bladder', 'lark vomit']
```

Первый пример — это список из четырёх целых чисел. Второй — список из трёх строк. Элементы списка не обязаны быть одного типа. Следующий список содержит строку, число с *плавающей точкой* (floating point number), целое число и ещё один список:


```python
['spam', 2.0, 5, [10, 20]]
```

Список, содержащий другой список, называется *вложенным* (nested).

Список, который не содержит элементов, называется *пустым списком* (empty list); его можно создать с помощью пустых квадратных скобок — `[]`.

Как и можно ожидать, значения списка можно присваивать переменным:

```python
cheeses = ['Cheddar', 'Edam', 'Gouda']
numbers = [17, 123]
empty = []
print(cheeses, numbers, empty)
```

### **Списки являются изменяемыми**

Списки являются *изменяемыми* (mutable), то есть их содержимое можно изменять после создания.

Синтаксис доступа к элементам списка такой же, как и при доступе к символам строки: используется оператор квадратных скобок. Выражение внутри скобок задаёт *индекс* (index). Помните, что индексация начинается с 0:


```python
print(cheeses[0])
```

В отличие от строк, списки являются изменяемыми, поскольку можно изменять порядок элементов в списке или переназначать отдельные элементы. Когда оператор квадратных скобок используется в левой части операции присваивания, он указывает на элемент списка, которому будет присвоено новое значение.


```python
numbers = [17, 123]
numbers[1] = 5
print(numbers)
```

Первый элемент списка <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">numbers</span>, который раньше был равен 123, теперь равен 5.

Список можно рассматривать как соответствие между индексами и элементами. Это соответствие называется *отображением* (mapping): каждый индекс «указывает на» один из элементов.

Индексы в списках работают так же, как и индексы в строках:

* В качестве индекса можно использовать любое целочисленное выражение.
* Если попытаться прочитать или изменить элемент, которого не существует, возникает ошибка <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">IndexError</span>.
* Если индекс имеет отрицательное значение, отсчёт ведётся с конца списка.

Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">in</span> также работает со списками.



```python
cheeses = ['Cheddar', 'Edam', 'Gouda']
print('Edam' in cheeses, 'Brie' in cheeses)
```

### **Обход списка**

Наиболее распространённый способ выполнить *обход* (traverse) элементов списка — использовать цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>. Синтаксис такой же, как и для строк:


```python
for cheese in cheeses:
    print(cheese)
```

Это хорошо работает, если требуется только читать элементы списка. Однако если нужно изменять или обновлять элементы, необходимы индексы. Распространённый способ — использовать функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">range</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len</span> вместе:


```python
for i in range(len(numbers)):
    numbers[i] = numbers[i] * 2
    print(numbers[i])
```

Этот цикл выполняет обход списка и обновляет каждый элемент. Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len</span> возвращает количество элементов в списке. Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">range</span> возвращает последовательность индексов от 0 до n − 1, где n — длина списка. При каждой итерации переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">i</span> получает индекс следующего элемента. Оператор присваивания в теле цикла использует <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">i</span> для чтения старого значения элемента и присваивания нового значения.

Цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>, выполняемый по пустому списку, ни разу не выполнит тело цикла:


```python
for x in empty:
    print('This never happens.')
```

Хотя список может содержать другой список, вложенный список всё равно считается одним элементом. Длина этого списка равна четырём:


```python
['spam', 1, ['Brie', 'Roquefort', 'Pol le Veq'], [1, 2, 3]]
```

### **Операции со списками**

Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">+</span> выполняет *конкатенацию* (concatenation) списков:


```python
a = [1, 2, 3]
b = [4, 5, 6]
c = a + b
print(c)
```

Аналогично, оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">*</span> повторяет список заданное количество раз:


```python
print([0] * 4)
print([1, 2, 3] * 3)
```

В первом примере список повторяется четыре раза. Во втором примере список повторяется три раза.


### **Срезы списков**

Оператор среза также работает со списками:



```python
t = ['a', 'b', 'c', 'd', 'e', 'f']
print(t[1:3])
print(t[:4])
print(t[3:])
```

Если опустить первый индекс, срез начинается с начала списка. Если опустить второй, срез продолжается до конца. Таким образом, если опустить оба индекса, срез будет копией всего списка.


```python
t[:]
```

Поскольку списки являются изменяемыми, часто полезно создать копию перед выполнением операций, которые изменяют список.

Оператор среза в левой части присваивания позволяет обновлять сразу несколько элементов:


```python
t = ['a', 'b', 'c', 'd', 'e', 'f']
t[1:3] = ['x', 'y']
print(t)
```

### **Методы списков**

Язык Python предоставляет методы, которые работают со списками. Например, метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">append</span> добавляет новый элемент в конец списка:


```python
t = ['a', 'b', 'c']
t.append('d')
print(t)
```

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">extend</span> принимает список в качестве аргумента и добавляет все его элементы в конец списка:


```python
t1 = ['a', 'b', 'c']
t2 = ['d', 'e']
t1.extend(t2)
print(t1)
```

В этом примере список <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">t2</span> остаётся неизменным.

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sort</span> упорядочивает элементы списка по возрастанию:


```python
t = ['d', 'c', 'e', 'b', 'a']
t.sort()
print(t)
```

Большинство методов списков являются функциями без возвращаемого значения: они изменяют список и возвращают <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span>. Если по ошибке написать <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">t = t.sort()</span>, результат вас разочарует.


### **Удаление элементов**

Существует несколько способов удалить элементы из списка. Если известен индекс элемента, который нужно удалить, можно использовать метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">pop</span>:


```python
t = ['a', 'b', 'c']
x = t.pop(1)
print(t)
print(x)
```

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">pop</span> изменяет список и возвращает удалённый элемент. Если не указать индекс, он удаляет и возвращает последний элемент.

Если удалённое значение не требуется, можно использовать оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">del</span>:


```python
t = ['a', 'b', 'c']
t.remove('b')
print(t)
```

Если известен элемент, который нужно удалить (но не его индекс), можно использовать метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">remove</span>:


```python
t = ['a', 'b', 'c']
t.remove('b')
print(t)
```

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">remove</span> возвращает <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span>.

Чтобы удалить несколько элементов, можно использовать оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">del</span> со срезом:


```python
t = ['a', 'b', 'c', 'd', 'e', 'f']
del t[1:5]
print(t)
```

Как обычно, срез выбирает все элементы до второго индекса, не включая его.


### **Списки и функции**

Существует ряд встроенных функций, которые можно применять к спискам и которые позволяют быстро обрабатывать список без необходимости писать собственные циклы:


```python
nums = [3, 41, 12, 9, 74, 15]
print(len(nums))
print(max(nums))
print(min(nums))
print(sum(nums))
print(sum(nums)/len(nums))
```

Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sum()</span> работает только тогда, когда элементы списка являются числами. Другие функции (например, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">max()</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len()</span> и т. д.) работают со списками строк и другими типами, элементы которых можно сравнивать.

Можно переписать ранее рассмотренную программу, вычисляющую среднее значение чисел, введённых пользователем, используя список.

Сначала программа для вычисления среднего без использования списка:


```python
total = 0
count = 0
while (True):
    inp = input('Enter a number: ')
    if inp == 'done': break
    value = float(inp)
    total = total + value
    count = count + 1

average = total / count
print('Average:', average)
```

В этой программе используются переменные <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">count</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">total</span>, чтобы хранить количество введённых пользователем чисел и их накопленную сумму по мере того, как программа многократно запрашивает ввод числа.

Можно просто сохранять каждое введённое пользователем число и затем использовать встроенные функции для вычисления суммы и количества в конце.


```python
numlist = list()
while (True):
    inp = input('Enter a number: ')
    if inp == 'done': break
    value = float(inp)
    numlist.append(value)

average = sum(numlist) / len(numlist)
print('Average:', average)
```

Перед началом цикла создаётся пустой список, а затем при каждом вводе числа оно добавляется в список с помощью метода append. В конце программы вычисляется сумма чисел в списке и делится на количество элементов списка, чтобы получить среднее значение.



### **Списки и строки**

Строка — это последовательность символов, а список — это последовательность значений, однако список символов — это не то же самое, что строка. Чтобы преобразовать строку в список символов, можно использовать функцию <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">list</span>:


```python
s = 'spam'
t = list(s)
print(t)
```

Поскольку <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">list</span> — это имя встроенной функции, его следует избегать в качестве имени переменной. Также часто избегают использования буквы «l», поскольку она слишком похожа на цифру «1». Поэтому в примерах используется имя переменной «t».

Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">list</span> разбивает строку на отдельные символы. Если требуется разбить строку на слова, можно использовать метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">split</span>:


```python
s = 'pining for the fjords'
t = s.split()
print(t)
print(t[2])
```

После того как строка разбита на список слов с помощью метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">split</span>, можно использовать оператор индексации (квадратные скобки), чтобы получить доступ к определённому слову в списке.

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">split</span> можно вызвать с необязательным аргументом — разделителем, который задаёт символы, используемые как границы слов. В следующем примере в качестве разделителя используется дефис:


```python
s = 'spam-spam-spam'
delimiter = '-'
s.split(delimiter)
```

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">join</span> является обратным по отношению к <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">split</span>. Он принимает список строк и объединяет его элементы в одну строку. Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">join</span> относится к строкам, поэтому его нужно вызывать у строки-разделителя, передавая список в качестве аргумента:


```python
t = ['pining', 'for', 'the', 'fjords']
delimiter = ' '
delimiter.join(t)
```

В данном случае разделителем является пробел, поэтому <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">join</span> вставляет пробел между словами. Чтобы объединить строки без пробелов, можно использовать пустую строку («») в качестве разделителя.


### **Разбор строк**

Обычно при чтении файла требуется не просто выводить строки целиком, а выполнять над ними дополнительные операции. Часто необходимо находить «интересные строки», а затем разбирать строку, чтобы извлечь из неё нужную часть. Например, если требуется вывести день недели из тех строк, которые начинаются с «From»?


From codestephen.marquard@uct.ac.zа Sat Jan  5 09:14:16 2008

Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">split</span> очень эффективен при решении подобных задач. Можно написать небольшую программу, которая ищет строки, начинающиеся с «From», разбивает эти строки и затем выводит третье слово в строке:


```python
fhand = open('mbox-short.txt')
for line in fhand:
    line = line.rstrip()
    if not line.startswith('From '): continue
    words = line.split()
    print(words[2])
```

Позже будут рассмотрены более продвинутые методы отбора нужных строк и их разбиения для извлечения точной информации, которая требуется.


### **Объекты и значения**

Если выполнить следующие операторы присваивания:


```python
a = 'banana'
b = 'banana'
```

Известно, что переменные <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">a</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">b</span> ссылаются на строку, но неизвестно, ссылаются ли они на один и тот же объект. Возможны два варианта:


Переменные и объекты



В одном случае переменные <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">a</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">b</span> ссылаются на два разных объекта с одинаковым значением. Во втором случае они ссылаются на один и тот же объект.

Чтобы проверить, ссылаются ли две переменные на один и тот же объект, можно использовать оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">is</span>.


```python
a = 'banana'
b = 'banana'
a is b
```

В этом примере Python создаёт только один строковый объект, и обе переменные <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">a</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">b</span> ссылаются на него.

Однако при создании двух списков создаются два отдельных объекта:


```python
a = [1, 2, 3]
b = [1, 2, 3]
a is b
```

В этом случае говорят, что два списка эквивалентны, поскольку содержат одинаковые элементы, но не идентичны, так как это разные объекты. Если два объекта идентичны, то они также эквивалентны, однако обратное не всегда верно: эквивалентные объекты не обязательно идентичны.

До этого момента термины «объект» и «значение» использовались как взаимозаменяемые, однако точнее говорить, что объект имеет значение. Если выполнить <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">a = [1,2,3]</span>, то переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">a</span> ссылается на объект списка, значением которого является определённая последовательность элементов. Если другой список содержит те же элементы, говорят, что он имеет то же значение.


### **Алиасинг**

Если переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">a</span> ссылается на объект и выполнить присваивание <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">b = a</span>, то обе переменные будут ссылаться на один и тот же объект:


```python
a = [1, 2, 3]
b = a
b is a
```

Связь переменной с объектом называется *ссылкой*. В этом примере существует две ссылки на один и тот же объект.

Объект, на который ссылаются более одной переменной, имеет несколько имён, поэтому говорят, что он *алиасирован*.

Если алиасированный объект является изменяемым, изменения, выполненные через одно имя, будут влиять и на другое:



```python
b[0] = 17
print(a)
```

Хотя такое поведение может быть полезным, оно подвержено ошибкам. В общем случае безопаснее избегать алиасинга при работе с изменяемыми объектами.

Для неизменяемых объектов, таких как строки, алиасинг не представляет серьёзной проблемы. В этом примере:


```python
a = 'banana'
b = 'banana'
```

почти никогда не имеет значения, ссылаются ли переменные <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">a</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">b</span> на одну и ту же строку или нет.


### **Списки как аргументы функций**

Когда список передаётся в функцию, функция получает ссылку на этот список. Если функция изменяет параметр-список, вызывающая сторона увидит это изменение. Например, функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">delete_head</span> удаляет первый элемент списка:


```python
def delete_head(t):
    del t[0]
```

Вот как она используется:


```python
letters = ['a', 'b', 'c']
delete_head(letters)
print(letters)
```

Параметр <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">t</span> и переменная <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">letters</span> являются алиасами одного и того же объекта.

Важно различать операции, которые изменяют списки, и операции, которые создают новые списки. Например, метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">append</span> изменяет список, а оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">+</span> создаёт новый список:


```python
t1 = [1, 2]
t2 = t1.append(3)
print(t1)
print(t2)

t3 = t1 + [3]
print(t3)
t1 is t3
```

Это различие важно при написании функций, которые должны изменять списки. Например, следующая функция не удаляет первый элемент списка:


```python
def bad_delete_head(t):
    t = t[1:]              # WRONG!
```

Оператор среза создаёт новый список, а операция присваивания делает переменную <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">t</span> ссылающейся на него, но это никак не влияет на список, который был передан в качестве аргумента.

Альтернативный подход — написать функцию, которая создаёт и возвращает новый список. Например, функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">tail</span> возвращает все элементы списка, кроме первого:


```python
def tail(t):
    return t[1:]
```

Эта функция не изменяет исходный список. Вот как она используется:


```python
letters = ['a', 'b', 'c']
rest = tail(letters)
print(rest)
```

**Упражнение 1:** Напишите функцию с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">chop</span>, которая принимает список, изменяет его, удаляя первый и последний элементы, и возвращает <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span>. Затем напишите функцию с именем <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">middle</span>, которая принимает список и возвращает новый список, содержащий все элементы, кроме первого и последнего.


```python
def chop(lst):
    lst[:] = lst[1:-1]
    return None


def middle(lst):
    new_list = lst[1:-1]
    return new_list


my_list = [1, 2, 3, 4, 5]

my_new_list = middle(my_list)
chop(my_list)

print("After chop:", my_list)
print("Middle:", my_new_list)
```

### **Отладка**

Неосторожное использование списков (и других изменяемых объектов) может привести к длительной отладке. Ниже приведены распространённые ошибки и способы их избежать:

1) Не забывайте, что большинство методов списков изменяют аргумент и возвращают <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span>. Это противоположно методам строк, которые возвращают новую строку и не изменяют исходную.

Если вы привыкли писать код для строк следующим образом:


```python
word = word.strip()
```

Возникает соблазн писать код для списков аналогичным образом:


```python
t = t.sort()           # WRONG!
```

Поскольку метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sort</span> возвращает <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">None</span>, следующая операция, которую вы попытаетесь выполнить с переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">t</span>, скорее всего, завершится ошибкой.

Прежде чем использовать методы и операторы списков, следует внимательно читать документацию, а затем проверять их в интерактивном режиме. Методы и операторы, которые списки разделяют с другими последовательностями (например, строками), описаны здесь:

docs.python.org/library/stdtypes.html#common-sequence-operations

Методы и операторы, которые применимы только к изменяемым последовательностям, описаны здесь:

docs.python.org/library/stdtypes.html#mutable-sequence-types

2) Придерживайтесь одного подхода и используйте его последовательно.

Часть проблемы со списками заключается в том, что существует слишком много способов выполнить одну и ту же операцию. Например, чтобы удалить элемент из списка, можно использовать методы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">pop</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">remove</span>, оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">del</span> или даже присваивание со срезом.

Чтобы добавить элемент, можно использовать метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">append</span> или оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">+</span>. Но не забывайте, что корректными являются следующие варианты:


```python
t.append(x)
t = t + [x]
```

А следующие варианты являются неправильными:


```python
t.append([x])          # WRONG!
t = t.append(x)        # WRONG!
t + [x]                # WRONG!
t = t + x              # WRONG!
```

Попробуйте выполнить каждый из этих примеров в интерактивном режиме, чтобы убедиться, что вы понимаете, что они делают. Обратите внимание, что только последний пример вызывает ошибку выполнения; остальные три допустимы, но работают не так, как ожидается.

3) Создавайте копии, чтобы избежать алиасинга.

Если требуется использовать метод, такой как <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sort</span>, который изменяет аргумент, но при этом необходимо сохранить исходный список, можно создать его копию.


```python
orig = t[:]
t.sort()
```

В этом примере можно также использовать встроенную функцию <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sorted</span>, которая возвращает новый отсортированный список и не изменяет исходный. Однако в этом случае не следует использовать <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">sorted</span> в качестве имени переменной.

Вернёмся к программе, которая ищет день недели в строках файла, начинающихся с «From»:

Можно долго смотреть на код и пытаться понять проблему самостоятельно или обратиться за помощью, но более быстрый и разумный подход — добавить оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">print</span>. Лучше всего добавить его непосредственно перед строкой, в которой программа завершилась с ошибкой, и вывести данные, которые, по-видимому, вызывают сбой.

Такой подход может породить много строк вывода, но по крайней мере сразу даст представление о сути проблемы. Поэтому перед пятой строкой добавляется вывод переменной <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">words</span>. В начало строки даже добавляется префикс "Debug:", чтобы отделить обычный вывод программы от отладочного вывода.


```python
fhand = open('mbox-short.txt')
count = 0
for line in fhand:
    words = line.split()
    # print('Debug:', words)
    if len(words) == 0 : continue
    if words[0] != 'From' : continue
    print(words[2])
```

Когда программа запускается, на экране быстро прокручивается большой объём вывода, но в конце видны отладочный вывод и трассировка ошибки, поэтому становится понятно, что произошло непосредственно перед её возникновением.

Каждая отладочная строка выводит список слов, который получается после разбиения строки на слова с помощью метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">split</span>. В момент сбоя программы этот список слов оказывается пустым: <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">[]</span>. Если открыть файл в текстовом редакторе и посмотреть его содержимое в этом месте, оно будет выглядеть следующим образом:

Ошибка возникает, когда программа встречает пустую строку. Разумеется, в пустой строке «ноль слов». Почему это не было учтено при написании кода? Когда программа пытается проверить первое слово через <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">word[0]</span>, чтобы сравнить его с «From», возникает ошибка «index out of range».

Именно здесь уместно добавить код по шаблону «стража», чтобы не проверять первое слово, если его вообще нет. Защитить этот код можно разными способами; в данном случае будет выбран вариант, при котором сначала проверяется количество слов, и только потом — первое слово:


```python
fhand = open('mbox-short.txt')
count = 0
for line in fhand:
    words = line.split()
    # print('Debug:', words)
    if len(words) == 0 : continue
    if words[0] != 'From' : continue
    print(words[2])
```

Сначала оператор отладочного вывода не удалили, а закомментировали, на случай если после внесённых изменений программу снова придётся отлаживать. Затем был добавлен оператор по шаблону «стража», который проверяет, не равно ли количество слов нулю, и если это так, то с помощью оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">continue</span> выполняется переход к следующей строке файла.

Можно считать, что два оператора <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">continue</span> помогают постепенно уточнять множество строк, которые представляют для нас интерес и которые нужно обрабатывать дальше. Строка, в которой нет слов, для нас неинтересна, поэтому выполняется переход к следующей строке. Строка, у которой первым словом не является «From», тоже не представляет интереса, поэтому она пропускается.

Программа в изменённом виде выполняется успешно, поэтому может показаться, что она написана правильно. Оператор по шаблону «стража» действительно гарантирует, что обращение к <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">words[0]</span> больше не приведёт к ошибке, но, возможно, этого всё ещё недостаточно. При программировании всегда нужно задаваться вопросом: «Что здесь может пойти не так?»

### **Глоссарий**

**Алиасинг (aliasing)**

Ситуация, при которой две или более переменные ссылаются на один и тот же объект.

**Разделитель (delimiter)**

Символ или строка, используемые для указания мест, в которых строка должна быть разбита.

**Элемент (element)**

Одно из значений в списке (или другой последовательности); также называется элементом списка.

**Эквивалентность (equivalent)**

Наличие одинакового значения.

**Индекс (index)**

Целочисленное значение, указывающее на элемент в списке.

**Идентичность (identical)**

Совпадение объектов (что подразумевает эквивалентность).

**Список (list)**

Последовательность значений.

**Обход списка (list traversal)**

Последовательный доступ к каждому элементу списка.

**Вложенный список (nested list)**

Список, являющийся элементом другого списка.

**Объект (object)**

Сущность, на которую может ссылаться переменная. Объект имеет тип и значение.

**Ссылка (reference)**

Связь между переменной и её значением.

```python
file_name = "romeo.txt"
unique_words = []

try:
    with open(file_name, "r") as fhand:
        for line in fhand:
            words = line.split()

            for word in words:
                if word not in unique_words:
                    unique_words.append(word)

    unique_words.sort()

    print(unique_words)

except FileNotFoundError:
    print(f"File {file_name} not found.")
```

**Упражнение 5:** Минималистичный почтовый клиент

MBOX (mail box) — это популярный формат файла для хранения и обмена коллекциями электронных писем. Он использовался в ранних почтовых серверах и настольных приложениях. Не вдаваясь в детали: MBOX представляет собой текстовый файл, в котором письма записаны последовательно. Письма разделяются специальной строкой, начинающейся с «From» (обратите внимание на пробел). Важно: строки, начинающиеся с «From:» (с двоеточием), описывают само письмо и не являются разделителями. Представьте, что вы разрабатываете минималистичное почтовое приложение, которое выводит адреса отправителей писем из входящих сообщений пользователя и подсчитывает их количество.

Напишите программу, которая читает данные почтового ящика и, находя строку, начинающуюся с «From», разбивает её на слова с помощью функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">split</span>. Нас интересует отправитель сообщения — это второе слово в строке «From».


From stephen.marquard@uct.ac.zа Sat Jan 5 09:14:16 2008

Необходимо разобрать строку «From» и вывести второе слово для каждой такой строки, а также подсчитать количество строк «From» (но не «From:») и вывести итоговое количество в конце. Ниже приведён пример корректного вывода (часть строк опущена):


```python
file_name = "mbox-short.txt"

try:
    fhand = open(file_name)
except:
    print("File cannot be opened:", file_name)
    quit()

count = 0

for line in fhand:
    line = line.rstrip()
    if not line.startswith("From "):
        continue
    words = line.split()
    print(words[1])
    count += 1

print(f"There were {count} lines in the file with From as the first word")
```

**Упражнение 6:**

Перепишите программу, которая запрашивает у пользователя список чисел и в конце выводит максимальное и минимальное значения после ввода «done». Напишите программу так, чтобы введённые пользователем числа сохранялись в список, а затем используйте функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">max()</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">min()</span> для вычисления максимального и минимального значений после завершения цикла.


```python
numbers = []

while True:
    user_input = input("Enter a number: ")

    if user_input == "done":
        break

    try:
        number = float(user_input)
        numbers.append(number)
    except ValueError:
        print("Invalid input")
        continue

if len(numbers) > 0:
    print(f"Maximum: {max(numbers)}")
    print(f"Minimum: {min(numbers)}")
else:
    print("No numbers were entered.")
```
