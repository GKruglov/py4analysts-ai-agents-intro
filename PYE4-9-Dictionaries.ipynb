## **Словари**

Словарь похож на список, но является более универсальной структурой данных. В списке в качестве индексов используются только целые числа, а в словаре индексами могут быть значения практически любого типа.

Словарь можно представить как отображение между набором индексов (которые называются **ключами**) и набором значений. Каждый ключ сопоставляется со значением. Соответствие между ключом и значением называется **парой ключ-значение**, или иногда — **элементом**.

В качестве примера построим словарь, сопоставляющий английские слова испанским, то есть и ключи, и значения в нём будут строками.

Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dict</span> создаёт новый словарь без элементов. Поскольку <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">dict</span> — это имя встроенной функции, не следует использовать его как имя переменной.


````python
eng2sp = dict()
print(eng2sp)
````

Фигурные скобки <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">{{}}</span> обозначают пустой словарь. Чтобы добавить элементы в словарь, можно использовать квадратные скобки:


````python
eng2sp['one'] = 'uno'
````

Эта строка создаёт элемент, сопоставляющий ключ <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">'one'</span> со значением «uno». Если снова вывести словарь, можно увидеть пару ключ-значение, разделённую двоеточием:


````python
print(eng2sp)
````

Такой формат вывода также можно использовать как формат ввода. Например, можно создать новый словарь сразу с тремя элементами:


````python
eng2sp = {'one': 'uno', 'two': 'dos', 'three': 'tres'}
print(eng2sp)
````

Начиная с Python 3.7, порядок пар ключ-значение совпадает с порядком их добавления, то есть словари теперь являются упорядоченными структурами.

Впрочем, это не так важно, потому что элементы словаря никогда не индексируются целыми числами. Вместо этого для получения соответствующего значения используется ключ:


````python
print(eng2sp['two'])
````

Ключ <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">'two'</span> всегда сопоставляется со значением «dos», поэтому порядок элементов не имеет значения.

Если указанный ключ отсутствует в словаре, возникает исключение:


````python
print(eng2sp['four'])
````

Функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">len</span> работает и со словарями; она возвращает количество пар ключ-значение:


````python
len(eng2sp)
````

Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">in</span> применим к словарям; он сообщает, присутствует ли элемент в качестве **ключа** в словаре (наличие элемента в качестве значения этим оператором не проверяется).


````python
print('one' in eng2sp)
print('uno' in eng2sp)
````

Чтобы проверить, присутствует ли элемент в качестве значения словаря, можно воспользоваться методом <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">values</span>, который возвращает значения в виде, который можно привести к списку, и далее применить оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">in</span>:


````python
vals = list(eng2sp.values())
'uno' in vals
````

Оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">in</span> использует разные алгоритмы для списков и для словарей. Для списков применяется линейный поиск: с увеличением длины списка время поиска растёт пропорционально количеству элементов. Для словарей в Python используется алгоритм **хеш-таблицы**, обладающий замечательным свойством: оператор <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">in</span> занимает примерно одинаковое время независимо от количества элементов в словаре. Подробнее о хеш-таблицах можно прочитать на [wikipedia.org/wiki/Hash_table](https://en.wikipedia.org/wiki/Hash_table).

### **Словарь как набор счётчиков**

Предположим, дана строка, и нужно подсчитать, сколько раз в ней встречается каждая буква. Сделать это можно несколькими способами:

- создать 26 переменных — по одной на каждую букву алфавита. Затем пройти по строке и при каждой букве увеличивать соответствующий счётчик, например с помощью цепочки условий;
- создать список из 26 элементов. Преобразовать каждый символ в число (с помощью встроенной функции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">ord</span>), использовать число как индекс списка и инкрементировать нужный счётчик;
- создать словарь, в котором ключами будут символы, а значениями — счётчики. При первой встрече символа добавлять элемент в словарь, далее — увеличивать значение существующего элемента.

Каждый вариант выполняет одно и то же вычисление, но реализуется по-разному.

**Реализация** — это конкретный способ выполнения вычисления; одни реализации лучше других. Например, преимущество реализации на словарях в том, что нам не нужно заранее знать, какие буквы встретятся в строке, — мы выделяем место только для тех букв, которые реально встречаются.

Возможный код выглядит так:


````python
word = 'brontosaurus'
d = dict()
for c in word:
    if c not in d:
        d[c] = 1
    else:
        d[c] = d[c] + 1
print(d)
````

По сути, мы вычисляем **гистограмму** — статистический термин для набора счётчиков (или частот).

Цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> проходит по строке. На каждой итерации, если символ <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">c</span> ещё не встречался в словаре, мы добавляем новый элемент с ключом <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">c</span> и начальным значением 1 (поскольку видим эту букву впервые). Если <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">c</span> уже есть в словаре, мы увеличиваем значение <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">d[c]</span> на 1.

Гистограмма показывает, что буквы «a» и «b» встречаются один раз, «o» — дважды и так далее.

У словарей есть метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">get</span>, принимающий ключ и значение по умолчанию. Если ключ присутствует в словаре, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">get</span> возвращает соответствующее значение, иначе — значение по умолчанию. Например:


````python
counts = { 'chuck' : 1 , 'annie' : 42, 'jan': 100}
print(counts.get('jan', 0))
print(counts.get('tim', 0))
````

С помощью <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">get</span> цикл для построения гистограммы можно записать более компактно. Так как метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">get</span> сам обрабатывает случай отсутствия ключа, четыре строки кода сокращаются до одной, а инструкция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span> становится ненужной.


````python
word = 'brontosaurus'
d = dict()
for c in word:
    d[c] = d.get(c,0) + 1
print(d)
````

Использование метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">get</span> для упрощения цикла подсчёта — очень распространённая «идиома» Python; в книге она встретится ещё много раз. Стоит уделить минуту и сравнить вариант с <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">if</span> и оператором <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">in</span> с вариантом, использующим метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">get</span>. Они делают одно и то же, но второй вариант лаконичнее.


### **Словари и файлы**

Одно из типичных применений словаря — подсчёт количества вхождений слов в файле с текстом. Начнём с очень простого файла, содержащего отрывок из «Ромео и Джульетты».

Для первой серии примеров возьмём сокращённую и упрощённую версию текста без знаков препинания. Позже мы перейдём к работе с текстом сцены со знаками препинания.

```
But soft what light through yonder window breaks
It is the east and Juliet is the sun
Arise fair sun and kill the envious moon
Who is already sick and pale with grief
```

Напишем программу на Python, которая читает файл построчно, разбивает каждую строку на список слов и затем, проходя по словам, подсчитывает их с помощью словаря.

У нас будет два вложенных цикла <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>. Внешний цикл читает строки файла, внутренний — перебирает слова конкретной строки. Это пример шаблона **вложенных циклов**: один цикл является **внешним**, а другой — **внутренним**.

Поскольку внутренний цикл за одну итерацию внешнего цикла выполняет все свои итерации целиком, внутренний цикл можно представить как «быстро» крутящийся, а внешний — как «медленно» крутящийся.

Сочетание двух вложенных циклов гарантирует, что мы подсчитаем каждое слово в каждой строке входного файла.


````python
fname = input('Enter the file name: ')
try:
    fhand = open(fname)
except:
    print('File cannot be opened:', fname)
    exit()

counts = dict()
for line in fhand:
    words = line.split()
    for word in words:
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] += 1

print(counts)
````

В блоке <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">else</span> мы используем более короткую запись для инкремента переменной. <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">counts[word] += 1</span> эквивалентно <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">counts[word] = counts[word] + 1</span>. Любой из способов может использоваться для изменения значения переменной на нужную величину. Аналогичные сокращения существуют для <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">-=</span>, <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">*=</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">/=</span>.

При запуске программы мы получаем «сырой» вывод всех счётчиков в неотсортированном хеш-порядке. Файл *romeo.txt* доступен по адресу [www.py4e.com/code3/romeo.txt](http://www.py4e.com/code3/romeo.txt).

Искать в таком выводе наиболее частые слова неудобно, поэтому в программу следует добавить ещё немного кода, чтобы получить более полезный результат.


### **Циклы и словари**

Если использовать словарь как последовательность в инструкции <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span>, цикл будет проходить по ключам словаря. Следующий цикл выводит каждый ключ и соответствующее ему значение:


````python
counts = { 'chuck' : 1 , 'annie' : 42, 'jan': 100}
for key in counts:
    print(key, counts[key])
````

Ключи выводятся в порядке добавления.

Этот шаблон можно использовать для реализации различных идиом цикла, рассмотренных ранее. Например, чтобы найти все элементы словаря со значением больше 10, можно написать следующий код:


````python
counts = { 'chuck' : 1 , 'annie' : 42, 'jan': 100}
for key in counts:
    if counts[key] > 10 :
        print(key, counts[key])
````

Цикл <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> проходит по **ключам** словаря, поэтому для получения **значения** по каждому ключу мы используем оператор индекса.

В результате выводятся только элементы со значением больше 10.

Чтобы вывести ключи в алфавитном порядке, можно получить список ключей с помощью метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">keys</span>, доступного у объектов-словарей, отсортировать этот список и, проходя по нему, выводить пары ключ-значение:


````python
counts = { 'chuck' : 1 , 'annie' : 42, 'jan': 100}
lst = list(counts.keys())
print(lst)
lst.sort()
print(lst)
for key in lst:
    print(key, counts[key])
````

Сначала выводится список ключей в порядке, полученном от метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">keys</span>, — ключи перечислены не по алфавиту. Затем следует вывод пар ключ-значение уже в алфавитном порядке, полученный в цикле <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">for</span> по отсортированному списку.


### **Продвинутый разбор текста**

В предыдущем примере с файлом *romeo.txt* мы максимально упростили файл, вручную убрав все знаки препинания. В настоящем тексте знаков препинания много, как показано ниже.

```
But, soft! what light through yonder window breaks?
It is the east, and Juliet is the sun.
Arise, fair sun, and kill the envious moon,
Who is already sick and pale with grief,
```

Поскольку функция <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">split</span> разделяет строку по пробелам и рассматривает слова как токены, разделённые пробелами, слова «soft!» и «soft» будут считаться **разными**, и для каждого из них в словаре появится отдельный элемент.

Кроме того, в файле есть заглавные буквы, и «who» и «Who» будут считаться разными словами с разными счётчиками.

Обе проблемы можно решить с помощью строкового метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">lower</span>, константы <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">string.punctuation</span> и метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">translate</span>. Метод <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">translate</span> — самый тонкий из перечисленных. Ниже приведена его документация:

<span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">line.translate(str.maketrans(fromstr, tostr, deletestr))</span>

Заменяет символы в <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">fromstr</span> соответствующими символами на тех же позициях в <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">tostr</span> и удаляет все символы, указанные в <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">deletestr</span>. <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">fromstr</span> и <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">tostr</span> могут быть пустыми строками, а параметр <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">deletestr</span> можно опустить.

Мы не будем указывать <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">tostr</span>, но используем параметр <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">deletestr</span> для удаления всех знаков препинания. При этом сам Python подскажет, какие символы считаются «знаками препинания»:


````python
import string
string.punctuation
````

В Python 2.0 параметры метода <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">translate</span> были другими.

Внесём в программу следующие изменения:


````python
import string

fname = input('Enter the file name: ')
try:
    fhand = open(fname)
except:
    print('File cannot be opened:', fname)
    exit()

counts = dict()
for line in fhand:
    line = line.rstrip()
    # First two parameters are empty strings
    line = line.translate(line.maketrans("", "", string.punctuation))
    line = line.lower()
    words = line.split()
    for word in words:
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] += 1

print(counts)
````

Часть освоения «искусства Python» или «мышления на Python» заключается в понимании того, что в Python часто уже есть встроенные возможности для многих типичных задач анализа данных. Проходить взглядом по такому выводу всё ещё неудобно, и с помощью Python можно получить именно то, что нам нужно. Но для этого понадобятся **кортежи** (tuples). К этому примеру мы вернёмся позже, после изучения кортежей.


### **Отладка**

При работе с большими наборами данных отладка «на глаз» с помощью вывода и ручной проверки становится неудобной. Вот несколько рекомендаций по отладке больших наборов данных:

**Уменьшайте размер данных.** Если программа читает текстовый файл, начните с первых 10 строк или с самого маленького примера, который получится найти. Файлы можно отредактировать вручную или (лучше) изменить программу, чтобы она читала только первые *n* строк. Если возникает ошибка, уменьшите *n* до минимального значения, при котором ошибка воспроизводится, а затем постепенно увеличивайте *n*, находя и исправляя ошибки.

**Печатайте сводки, а не данные целиком.** Например, выводите количество элементов в словаре или сумму элементов списка, а не весь набор данных.

**Проверяйте типы.** Одна из типичных причин ошибок времени выполнения — значение не того типа. Часто достаточно вывести тип значения.

**Пишите автоматические проверки.** Например, если вы вычисляете среднее списка чисел, можно проверить, что результат не превышает максимальный элемент и не меньше минимального. Такая проверка называется «проверкой на вменяемость» (sanity check), поскольку она выявляет «совершенно нелогичные» результаты.

**Сравнивайте результаты разных вычислений.** Если два независимых вычисления приводят к согласующимся результатам — это «проверка согласованности» (consistency check).

Время, потраченное на построение «лесов» для отладки, обычно окупается сокращением общего времени отладки.


### **Глоссарий**

**Словарь (dictionary)** — отображение ключей в соответствующие им значения.

**Гистограмма (histogram)** — набор счётчиков (или частот).

**Реализация (implementation)** — способ выполнения вычисления.

**Элемент (item)** — другое название пары ключ-значение.

**Ключ (key)** — объект, который появляется в словаре в составе пары ключ-значение; выступает первым элементом этой пары.

**Пара ключ-значение (key-value pair)** — представление сопоставления ключа и значения.

**Поиск (lookup)** — операция получения значения по ключу.

**Хеш-таблица (hash table)** — алгоритм, используемый функцией словаря Python.

**Вложенный цикл (nested loops)** — когда один цикл находится «внутри» другого. Внутренний цикл выполняется полностью за каждую итерацию внешнего.

**Значение (value)** — объект, который появляется в словаре в составе пары ключ-значение; выступает вторым элементом этой пары. Это более специфический смысл термина «значение», чем использованный ранее.

**Упражнение 2:** Напишите программу, которая классифицирует каждое почтовое сообщение по дню недели, когда оно было отправлено. Для этого ищите строки, начинающиеся с <span style="color: #c70000; background-color: #da7ec325; padding:2px 6px; border-radius:4px;">From</span>, затем берите третье слово и ведите счётчик для каждого дня недели. В конце программы выведите содержимое словаря (порядок не важен).

Пример строки:

```
From stephen.marquard@uct.ac.za Sat Jan  5 09:14:16 2008
```

Пример выполнения:

```
python dow.py
Enter a file name: mbox-short.txt
{'Fri': 20, 'Thu': 6, 'Sat': 1}
```


````python
day_counts = {}
try:
    with open("mbox-short.txt", "r") as file:
        for line in file:
            if line.startswith("From "):
                words = line.split()

                if len(words) > 3:
                    day = words[2]

                    day_counts[day] = day_counts.get(day, 0) + 1
except FileNotFoundError:
    print(f"File not found")
print(day_counts)
````

**Упражнение 4:** Дополните предыдущую программу так, чтобы определить, кто отправил больше всего сообщений. После того как все данные прочитаны и словарь построен, пройдите по словарю с помощью цикла поиска максимума (см. главу 5: «Циклы поиска максимума и минимума»), чтобы найти, от кого пришло наибольшее число сообщений, и выведите это число.

```
Enter a file name: mbox-short.txt
cwen@iupui.edu 5

Enter a file name: mbox.txt
zqian@umich.edu 195
```


````python
email_counts = dict()

try:
    file = open("mbox-short.txt", "r")
except:
    print("File can't be opened")

for line in file:
    if line.startswith("From "):
        words = line.split()
        if len(words) >= 2:
            email = words[1]
            email_counts[email] = email_counts.get(email, 0) + 1

max_email = None
max_count = None

for email, count in email_counts.items():
    if max_count is None or count > max_count:
        max_email = email
        max_count = count

print(max_email, max_count)
````
